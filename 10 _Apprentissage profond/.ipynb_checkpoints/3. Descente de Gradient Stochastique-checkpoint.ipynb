## Introduction  
Dans les deux premi√®res le√ßons, nous avons appris √† construire des r√©seaux enti√®rement connect√©s √† partir de piles de couches denses. Lorsqu'il est cr√©√©, tous les poids du r√©seau sont initialis√©s de mani√®re al√©atoire ‚Äî le r√©seau ne "**sait**" rien encore. Dans cette le√ßon, nous allons voir comment **entra√Æner un r√©seau de neurones**, c'est-√†-dire comment les r√©seaux de neurones apprennent.

Comme pour toutes les t√¢ches d'**apprentissage automatique**, nous commen√ßons avec un ensemble de **donn√©es d'entra√Ænement**. Chaque exemple dans les donn√©es d'entra√Ænement est compos√© de certaines caract√©ristiques (**les entr√©es**) ainsi que d'une cible attendue (**la sortie**). Entra√Æner le r√©seau signifie ajuster ses poids de mani√®re √† ce qu'il puisse transformer les caract√©ristiques en la cible. 

Par exemple, dans le jeu de donn√©es des 80 c√©r√©ales, nous voulons un r√©seau capable de prendre la teneur en '**sucre**', '**fibres**' et '**prot√©ines**' de chaque c√©r√©ale et de produire une pr√©diction pour les '**calories**' de cette c√©r√©ale. Si nous parvenons √† entra√Æner un r√©seau pour faire cela, ses poids doivent repr√©senter d'une certaine mani√®re la relation entre ces caract√©ristiques et cette cible, telle qu'elle est exprim√©e dans les donn√©es d'entra√Ænement.

En plus des donn√©es d'entra√Ænement, nous avons besoin de deux choses suppl√©mentaires :

1. Une **fonction de perte** qui mesure la qualit√© des pr√©dictions du r√©seau.
2. Un **optimiseur** qui peut indiquer au r√©seau comment modifier ses poids.

---
### **La Fonction de Perte**  
Nous avons vu comment concevoir une architecture pour un r√©seau, mais nous n'avons pas vu comment indiquer au r√©seau quel probl√®me r√©soudre. C'est le r√¥le de la fonction de perte.

**La fonction de perte mesure l'√©cart entre la vraie valeur de la cible et la valeur pr√©dite par le mod√®le.**

Diff√©rents probl√®mes n√©cessitent diff√©rentes fonctions de perte. Nous avons examin√© des **probl√®mes de r√©gression**, o√π la t√¢che consiste √† pr√©dire une valeur num√©rique ‚Äî comme les calories dans les 80 c√©r√©ales ou la note dans la qualit√© du vin rouge. D'autres t√¢ches de r√©gression pourraient √™tre la pr√©diction du prix d'une maison ou de l'efficacit√© √©nerg√©tique d'une voiture.

Une fonction de perte courante pour les probl√®mes de r√©gression est **l'erreur absolue moyenne (MAE)**. Pour chaque pr√©diction `y_pred`, la **MAE** mesure l'√©cart par rapport √† la vraie cible `y_true` par une diff√©rence absolue `abs(y_true - y_pred)`.

**La perte MAE totale sur un ensemble de donn√©es est la moyenne de toutes ces diff√©rences absolues.**

Un graphique repr√©sentant les barres d'erreur des points de donn√©es √† la courbe ajust√©e.  
L'erreur absolue moyenne est la longueur moyenne entre la courbe ajust√©e et les points de donn√©es.

Outre la MAE, d'autres fonctions de perte que vous pourriez rencontrer pour les probl√®mes de r√©gression sont l'**erreur quadratique moyenne (MSE)** ou la **perte de Huber** (toutes deux disponibles dans Keras).

Pendant l'entra√Ænement, le mod√®le utilisera la fonction de perte comme guide pour trouver les valeurs correctes de ses poids (**une perte plus faible est meilleure**). En d'autres termes, la fonction de perte indique au r√©seau son objectif.


---
## **L'Optimiseur - Descente de Gradient Stochastique (SGD)**  
Nous avons d√©crit le probl√®me que nous voulons que le r√©seau r√©solve, mais maintenant nous devons dire comment le r√©soudre. C'est le r√¥le de l'optimiseur. **L'optimiseur est un algorithme qui ajuste les poids pour minimiser la perte**.

Pratiquement tous les algorithmes d'optimisation utilis√©s en apprentissage profond appartiennent √† une famille appel√©e **descente de gradient stochastique**. Ce sont des algorithmes **it√©ratifs** qui entra√Ænent un r√©seau par √©tapes. Une √©tape d'entra√Ænement se d√©roule comme suit :

- 1. √âchantillonner quelques donn√©es d'entra√Ænement et les faire passer √† travers le r√©seau pour faire des pr√©dictions.
- 2. Mesurer la perte entre les pr√©dictions et les vraies valeurs.
- 3. Enfin, ajuster les poids dans une direction qui r√©duit la perte.

Ensuite, r√©p√©tez ce processus jusqu'√† ce que la perte soit aussi petite que vous le souhaitez (ou jusqu'√† ce qu'elle ne diminue plus).

Ajuster une ligne lot par lot. La perte diminue et les poids approchent leurs vraies valeurs.  
Entra√Æner un r√©seau de neurones avec la descente de gradient stochastique.

Chaque √©chantillon de donn√©es d'entra√Ænement dans une **it√©ration** est appel√© un **minibatch** (ou souvent simplement "**batch**"), tandis qu'un cycle complet des donn√©es d'entra√Ænement est appel√© **une √©poque**. Le nombre d'√©poques pour lesquelles vous entra√Ænez le r√©seau correspond au nombre de fois que le r√©seau verra chaque exemple d'entra√Ænement.

L'animation montre le mod√®le lin√©aire de la le√ßon 1 entra√Æn√© avec SGD. Les points rouges p√¢les repr√©sentent l'ensemble de l'ensemble d'entra√Ænement, tandis que les points rouges pleins sont les minibatchs. Chaque fois que SGD voit un nouveau minibatch, il d√©place les poids (w la pente et b l'ordonn√©e √† l'origine) vers leurs valeurs correctes pour ce lot. Lot apr√®s lot, la ligne finit par converger vers son meilleur ajustement. Vous pouvez voir que la perte diminue √† mesure que les poids se rapprochent de leurs vraies valeurs.


---
## **Taux d'Apprentissage et Taille des Lots** 
Remarquez que la ligne ne fait qu'un petit d√©calage dans la direction de chaque lot (au lieu de se d√©placer compl√®tement). La taille de ces d√©calages est d√©termin√©e par le taux d'apprentissage. Un **taux d'apprentissage** plus petit signifie que le r√©seau doit voir plus de minibatchs avant que ses poids ne convergent vers leurs meilleures valeurs.

Le taux d'apprentissage et la taille des minibatchs sont les deux param√®tres qui ont le plus grand effet sur la fa√ßon dont l'entra√Ænement SGD se d√©roule. Leur interaction est souvent subtile et le bon choix pour ces param√®tres n'est pas toujours √©vident. (Nous explorerons ces effets dans l'exercice.)

Heureusement, pour la plupart des travaux, il ne sera pas n√©cessaire de faire une recherche approfondie des hyperparam√®tres pour obtenir des r√©sultats satisfaisants. 
**Adam** est un algorithme **SGD** qui a un taux d'apprentissage adaptatif, ce qui le rend adapt√© √† la plupart des probl√®mes sans r√©glage des param√®tres (il est en quelque sorte "**auto-ajustable**"). **Adam est un excellent optimiseur polyvalent**.


## **Ajouter la Fonction de Perte et l'Optimiseur**  
Apr√®s avoir d√©fini un mod√®le, vous pouvez ajouter une fonction de perte et un optimiseur avec la m√©thode `compile` du mod√®le :


```
model.compile(
    optimizer="adam",
    loss="mae",

```

Remarquez que nous pouvons sp√©cifier la perte et l'optimiseur simplement avec une cha√Æne de caract√®res. Vous pouvez √©galement y acc√©der directement via l'**API Keras** ‚Äî si vous souhaitez ajuster des param√®tres, par exemple ‚Äî mais pour nous, les valeurs par d√©faut fonctionneront bien.

## **Que Signifie le Nom ?** 
Le **gradient** est un vecteur qui nous indique dans quelle direction les poids doivent aller. Plus pr√©cis√©ment, il nous indique comment modifier les poids pour que la perte change le plus rapidement. Nous appelons notre processus "**descente de gradient**" car il utilise le gradient pour descendre la courbe de perte vers un minimum. "**Stochastique**" signifie "d√©termin√© par le hasard." Notre entra√Ænement est stochastique car les minibatchs sont des √©chantillons al√©atoires de l'ensemble de donn√©es. Et c'est pourquoi on l'appelle **SGD** !

# **Exemple - Qualit√© du Vin Rouge**  
Maintenant, nous savons tout ce dont nous avons besoin pour commencer √† entra√Æner des mod√®les d'apprentissage profond. Voyons cela en action ! Nous allons utiliser le jeu de donn√©es sur la qualit√© du vin rouge.

Ce jeu de donn√©es contient des mesures physico-chimiques d'environ 1600 vins rouges portugais. Il inclut √©galement une note de qualit√© pour chaque vin, issue de tests de d√©gustation √† l'aveugle. **√Ä quel point pouvons-nous pr√©dire la qualit√© per√ßue d'un vin √† partir de ces mesures ?**

Nous avons int√©gr√© toute la pr√©paration des donn√©es dans la cellule suivante. Ce n'est pas essentiel pour ce qui suit, alors n'h√©sitez pas √† la sauter. Une chose que vous pourriez noter pour l'instant est que nous avons redimensionn√© chaque caract√©ristique pour qu'elle se situe dans l'intervalle [0, 1]. Comme nous en discuterons davantage dans la le√ßon 5, les r√©seaux de neurones ont tendance √† mieux performer lorsque leurs entr√©es sont sur une √©chelle commune.


````python
import pandas as pd
from IPython.display import display

red_wine = pd.read_csv('red-wine.csv')

# Cr√©er des ensembles d'entra√Ænement et de validation
df_train = red_wine.sample(frac=0.7, random_state=0)
df_valid = red_wine.drop(df_train.index)
display(df_train.head(4))

# Redimensionner √† [0, 1]
max_ = df_train.max(axis=0)
min_ = df_train.min(axis=0)
df_train = (df_train - min_) / (max_ - min_)
df_valid = (df_valid - min_) / (max_ - min_)

# S√©parer les caract√©ristiques et la cible
X_train = df_train.drop('quality', axis=1)
X_valid = df_valid.drop('quality', axis=1)
y_train = df_train['quality']
y_valid = df_valid['quality']
````

**Combien d'entr√©es ce r√©seau devrait-il avoir ?** Nous pouvons le d√©couvrir en regardant le nombre de colonnes dans la matrice de donn√©es. Assurez-vous de ne pas inclure la cible ('quality') ici ‚Äî uniquement les caract√©ristiques d'entr√©e.


````python
print(X_train.shape)
````

Onze colonnes signifient onze entr√©es.

Nous avons choisi un r√©seau √† trois couches avec plus de 1500 neurones. Ce r√©seau devrait √™tre capable d'apprendre des relations assez complexes dans les donn√©es.


````python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Input(shape=[11]), 
    layers.Dense(512, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(1),
])
````

D√©terminer l'**architecture** de votre mod√®le devrait faire partie d'un processus. Commencez simplement et utilisez la **perte de validation** comme guide. Vous en apprendrez plus sur le d√©veloppement de mod√®les dans les exercices.

Apr√®s avoir d√©fini le mod√®le, nous **compilons l'optimiseur et la fonction de perte**.

````python
model.compile(
    optimizer='adam',
    loss='mae',
)
````

Maintenant, nous sommes pr√™ts √† commencer l'entra√Ænement ! Nous avons dit √† **Keras** de donner √† l'optimiseur **256 lignes** des donn√©es d'entra√Ænement √† la fois (le `batch_size`) et de le faire **10 fois** √† travers l'ensemble de donn√©es (les `epochs`).


````python
history = model.fit(
    X_train, y_train,
    validation_data=(X_valid, y_valid),
    batch_size=256,
    epochs=10,
)
````

Vous pouvez voir que Keras vous tiendra inform√© de la perte pendant que le mod√®le s'entra√Æne.

Souvent, une meilleure fa√ßon de visualiser la perte est de la tracer. La m√©thode `fit` conserve en fait un enregistrement de la perte produite pendant l'entra√Ænement dans un objet `History`. Nous allons convertir les donn√©es en un dataframe Pandas, ce qui facilite le tra√ßage.


````python
import pandas as pd

# Convertir l'historique d'entra√Ænement en dataframe
history_df = pd.DataFrame(history.history)
# Utiliser la m√©thode de tra√ßage native de Pandas
history_df['loss'].plot();
````

Remarquez comment la perte se stabilise √† mesure que les √©poques passent. Lorsque la courbe de perte devient horizontale comme cela, cela signifie que le mod√®le a appris tout ce qu'il pouvait et qu'il n'y a aucune raison de continuer pour des √©poques suppl√©mentaires.

### **Pr√©sentation du dataset Fuel Economy üöó**  
Dans ce dataset, votre objectif est de **pr√©dire la consommation de carburant d'un v√©hicule** (*Fuel Economy - FE*) √† partir de plusieurs caract√©ristiques, comme **le type de moteur, l'ann√©e de fabrication**, etc.  

üìå **Chargement des donn√©es :** 

````python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer, make_column_selector
from sklearn.model_selection import train_test_split

# Chargement des donn√©es
fuel = pd.read_csv('fuel.csv')

# Copie des donn√©es
X = fuel.copy()
y = X.pop('FE')  # Suppression de la colonne cible (FE) des caract√©ristiques

X.head()
````

üìå **Pr√©traitement des donn√©es :** 

````python
preprocessor = make_column_transformer(
    (StandardScaler(), make_column_selector(dtype_include=np.number)),  # Normalisation des variables num√©riques
    (OneHotEncoder(sparse_output=False), make_column_selector(dtype_include=object)),  # Encodage des variables cat√©gorielles
)

X = preprocessor.fit_transform(X)
y = np.log(y)  # Transformation logarithmique de la cible (au lieu de la standardiser)

input_shape = [X.shape[1]]  # Taille des entr√©es
print("Input shape: {}".format(input_shape))  # Affichage de la dimension des entr√©es
````

üìå **Visualisation des donn√©es (facultatif) :**

````python
# Afficher les premi√®res lignes du dataset original
fuel.head()
````

````python
# Afficher les 10 premi√®res lignes des donn√©es pr√©trait√©es
pd.DataFrame(X[:10, :]).head()
````


---

### **D√©finition du mod√®le de r√©seau de neurones üß†**  
On d√©finit un **r√©seau de neurones** comportant **trois couches cach√©es** et une couche de sortie :  
- **2 couches cach√©es** de **128 neurones** avec activation **ReLU**  
- **1 couche cach√©e** de **64 neurones** avec activation **ReLU**  
- **1 couche de sortie** avec **1 neurone** (sans activation, car il s'agit d'un probl√®me de r√©gression)  

---

### üßê **R√©sum√©**  
| Couche | Nombre de neurones | Activation | Raison |
|--------|------------------|------------|--------|
| **Cach√©e 1** | 128 | ReLU | Capturer des relations complexes |
| **Cach√©e 2** | 128 | ReLU | Approfondir l'apprentissage |
| **Cach√©e 3** | 64  | ReLU | R√©duction progressive des dimensions |
| **Sortie** | 1 | Aucune | Pr√©dire un nombre r√©el (r√©gression) |

### **üí° Pourquoi pas plus ou moins de couches ?**  
- **Plus de couches** : Peut am√©liorer la performance, mais augmente le risque de **sur-apprentissage** et rend l'entra√Ænement plus lent.  
- **Moins de couches** : Peut ne pas capturer suffisamment les **relations complexes** entre les variables.  

üëâ **En pratique, ces hyperparam√®tres (nombre de couches, neurones, activation) sont souvent optimis√©s par validation crois√©e et ajustements successifs.** üîß

````python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Input(shape=input_shape),  # D√©finition explicite de l'entr√©e
    layers.Dense(128, activation='relu'),  # Couche cach√©e 1
    layers.Dense(128, activation='relu'),  # Couche cach√©e 2
    layers.Dense(64, activation='relu'),  # Couche cach√©e 3
    layers.Dense(1),  # Couche de sortie
])
````

Ce mod√®le est pr√™t √† √™tre entra√Æn√© pour pr√©dire la consommation de carburant des voitures ! üöóüí®

--- 

## Question 1) Ajouter la fonction de perte et l'optimiseur  
Avant d'entra√Æner le r√©seau, nous devons d√©finir **la fonction de perte (loss function) et l'optimiseur (optimizer)** que nous allons utiliser.  

En utilisant la m√©thode `.compile()` du mod√®le, ajoutez :  
- **L'optimiseur Adam** (*Adaptive Moment Estimation*), qui est un optimiseur avanc√© combinant les avantages de SGD et RMSprop.  
- **La fonction de perte MAE** (*Mean Absolute Error*), qui est couramment utilis√©e pour les probl√®mes de r√©gression car elle mesure l'√©cart moyen absolu entre les valeurs pr√©dites et r√©elles.  

````python
# D√©finition de la fonction de perte et de l'optimiseur
model.compile(optimizer="adam", loss="mae")
````

---

## Question 2) Entra√Æner le mod√®le

Une fois que vous avez d√©fini le mod√®le et que vous l'avez compil√© avec une fonction de perte et un optimiseur, vous √™tes pr√™t √† l'entra√Æner.  

Entra√Ænez le r√©seau pendant **200 √©poques** avec une **taille de batch de 128**. Les donn√©es d'entr√©e sont **X** et la cible est **y**.

Le choix de **200 √©poques** est fait pour s'assurer que le mod√®le a suffisamment de temps pour apprendre des donn√©es et am√©liorer ses performances.  

##### Pourquoi **200** et pas une autre valeur ?
1. **Assez d'it√©rations pour l'apprentissage**  
   - Au d√©but, le mod√®le ajuste rapidement ses poids.  
   - Apr√®s plusieurs √©poques, l'am√©lioration devient plus lente.  
   - 200 √©poques permettent d'explorer un bon √©quilibre entre **sous-apprentissage** (pas assez d'√©poques) et **surapprentissage** (trop d'√©poques).  

2. **Exp√©rience et empirisme**  
   - Pour des probl√®mes similaires en **r√©gression avec des r√©seaux de neurones**, 100 √† 200 √©poques sont souvent suffisantes.  
   - On peut ajuster ce nombre apr√®s observation des performances.  

3. **Observation du comportement du mod√®le**  
   - Si la **perte diminue toujours** apr√®s 200 √©poques, il faudra peut-√™tre augmenter le nombre d'√©poques.  
   - Si la **perte se stabilise bien avant**, on pourrait r√©duire le nombre d'√©poques pour **gagner du temps de calcul**.  

üí° **Id√©alement**, on utilise **"early stopping"**, qui arr√™te l'entra√Ænement automatiquement si la performance ne s'am√©liore plus apr√®s un certain nombre d'√©poques.

````python
# YOUR CODE HERE
history = model.fit(
    X, y,
    epochs=200,
    batch_size=128
)
````

La derni√®re √©tape consiste √† observer les courbes de perte et √† √©valuer l'entra√Ænement. Ex√©cutez la cellule ci-dessous pour afficher un graphique de la perte d'entra√Ænement.  


````python
import pandas as pd

history_df = pd.DataFrame(history.history)
# Commence le trac√© √† partir de l'√©poque 5. Vous pouvez modifier cette valeur pour une autre vue.
history_df.loc[5:, ['loss']].plot();
````

### √âvaluation de l'entra√Ænement

L'id√©e centrale ici est de se demander si la perte continuera de diminuer si vous entra√Ænez le mod√®le plus longtemps. En d'autres termes, si vous ajoutez plus d'√©poques d'entra√Ænement, est-ce que la perte (la fonction de co√ªt) continuera √† baisser ou s'atteindra-t-elle un minimum ? Cela d√©pend de l'√©volution de la courbe de la perte pendant l'entra√Ænement.

#### Sc√©narios :

1. **Si la courbe d'apprentissage s'est stabilis√©e** :
   - Si la perte a atteint un point o√π elle reste constante ou fluctue peu, alors **il n'y a g√©n√©ralement aucun avantage** √† entra√Æner le mod√®le plus longtemps. Cela signifie que le mod√®le a d√©j√† converg√© vers un minimum local et qu'il est peu probable que l'ajout de nouvelles √©poques am√©liore les r√©sultats.
   
2. **Si la perte continue de diminuer** :
   - Si la perte continue de baisser, cela signifie que le mod√®le est encore en train d'apprendre et pourrait b√©n√©ficier de plus d'√©poques d'entra√Ænement. Dans ce cas, l'entra√Ænement prolong√© pourrait effectivement **am√©liorer les performances** du mod√®le.

### L'impact du taux d'apprentissage (learning rate) et de la taille des lots (batch size)

Le **taux d'apprentissage** et la **taille des lots** influencent plusieurs aspects de l'entra√Ænement :

- **Le temps n√©cessaire pour entra√Æner le mod√®le** : Un grand taux d'apprentissage peut rendre l'entra√Ænement plus rapide, mais cela peut aussi rendre le processus plus instable.
  
- **Le bruit des courbes d'apprentissage** : Des petites tailles de lots augmentent le bruit (variabilit√©) des courbes d'apprentissage, car chaque lot ne repr√©sente qu'une petite fraction des donn√©es. √Ä l'inverse, de grands lots donnent des mises √† jour plus stables, mais peuvent rendre l'entra√Ænement plus lent.

- **La valeur minimale que peut atteindre la perte** : Un taux d'apprentissage trop √©lev√© peut entra√Æner une perte de qualit√© dans l'entra√Ænement, emp√™chant le mod√®le de converger vers un bon minimum.

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Fonction de descente de gradient stochastique
def sgd_step(w, b, x, y, learning_rate):
    # Calculer les gradients
    gradient_w = 2 * np.mean((w * x + b - y) * x)
    gradient_b = 2 * np.mean(w * x + b - y)
    
    # Mettre √† jour les poids
    w -= learning_rate * gradient_w
    b -= learning_rate * gradient_b
    return w, b

# Fonction d'animation de la descente de gradient stochastique
def animate_sgd(learning_rate, batch_size, num_examples, steps, true_w, true_b):
    # Cr√©er des donn√©es simul√©es
    X = np.random.rand(num_examples)
    Y = true_w * X + true_b + np.random.normal(scale=0.1, size=num_examples)

    # Initialiser les poids
    w, b = np.random.randn(), np.random.randn()

    fig, ax = plt.subplots()
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)

    # Line et point d'animation
    line, = ax.plot([], [], lw=2)
    point, = ax.plot([], [], 'ro')

    # Fonction d'initialisation
    def init():
        line.set_data([], [])
        point.set_data([], [])
        return line, point

    # Fonction de mise √† jour
    def update(frame):
        nonlocal w, b
        # Effectuer une √©tape de descente de gradient
        w, b = sgd_step(w, b, X[:batch_size], Y[:batch_size], learning_rate)
        # Mettre √† jour la ligne et les points
        line.set_data(X, w * X + b)
        point.set_data(w, b)
        return line, point

    # Cr√©er l'animation
    ani = FuncAnimation(fig, update, frames=steps, init_func=init, blit=True)
    return ani

# Liste des diff√©rentes combinaisons √† tester
param_combinations = [
    (0.05, 32, 256),
    (0.05, 2, 256),
    (0.05, 128, 256),
    (0.02, 32, 256),
    (0.2, 32, 256),
    (1.0, 32, 256),
    (0.9, 4096, 8192),
    (0.99, 4096, 8192)
]

# Tester chaque combinaison
for learning_rate, batch_size, num_examples in param_combinations:
    print(f"Testing with: Learning Rate = {learning_rate}, Batch Size = {batch_size}, Num Examples = {num_examples}")
    
    # Ex√©cution de l'animation avec les param√®tres actuels
    ani = animate_sgd(
        learning_rate=learning_rate,
        batch_size=batch_size,
        num_examples=num_examples,
        steps=50,  # total training steps (batches seen)
        true_w=3.0,  # the slope of the data
        true_b=2.0,  # the bias of the data
    )
    
    # Forcer l'affichage de l'animation
    plt.show()
````

## **Cr√©er la fonction animate_sgd**

````python
import numpy as np
import matplotlib.pyplot as plt

def animate_sgd(learning_rate=0.05, batch_size=32, num_examples=256, steps=50, true_w=3.0, true_b=2.0):
    # Initialisation des param√®tres
    w = np.random.randn()  # Poids initial
    b = np.random.randn()  # Biais initial
    X = np.random.randn(num_examples, 1)  # Donn√©es d'entr√©e
    y = X * true_w + true_b + np.random.randn(num_examples, 1)  # Donn√©es de sortie (avec bruit)

    # Variables pour stocker l'√©volution du mod√®le
    ws, bs = [], []

    for step in range(steps):
        # Calcul des gradients (descente de gradient)
        y_pred = X * w + b
        error = y_pred - y
        dw = 2 * np.mean(error * X)  # Gradient par rapport au poids
        db = 2 * np.mean(error)      # Gradient par rapport au biais

        # Mise √† jour des param√®tres
        w -= learning_rate * dw
        b -= learning_rate * db

        # Sauvegarder l'√©volution pour l'animation
        ws.append(w)
        bs.append(b)

    # Affichage de l'animation
    plt.plot(ws, label="Poids")
    plt.plot(bs, label="Biais")
    plt.legend()
    plt.xlabel('√âtape')
    plt.ylabel('Valeur')
    plt.title('Animation de la descente de gradient')
    plt.show()

# Appel de la fonction
animate_sgd()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# G√©n√©rer des donn√©es d'exemple
true_w = 3.0  # Pente r√©elle
true_b = 2.0  # Biais r√©el
num_examples = 256
X = np.random.randn(num_examples, 1)  # Donn√©es d'entr√©e
y = X * true_w + true_b + np.random.randn(num_examples, 1)  # Donn√©es de sortie avec bruit

# Initialisation des param√®tres du mod√®le
w = np.random.randn()
b = np.random.randn()

# Param√®tres de l'entra√Ænement
learning_rate = 0.1
steps = 100
history = []  # Pour stocker l'√©volution de la perte

# Fonction de calcul de la perte (MSE)
def compute_loss(X, y, w, b):
    y_pred = X * w + b
    return np.mean((y_pred - y) ** 2)

# Pr√©paration de l'animation
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.set_title('Fitted Line')
ax1.set_xlabel('X')
ax1.set_ylabel('y')
ax2.set_title('Training Loss')
ax2.set_xlabel('Steps')
ax2.set_ylabel('Loss')

line, = ax1.plot([], [], label='Line ajust√©e', color='red')
loss_line, = ax2.plot([], [], label='Training Loss', color='blue')
ax1.scatter(X, y, alpha=0.5)

# Fonction de mise √† jour de l'animation
def update(step):
    global w, b
    # Calcul des pr√©dictions et de l'erreur
    y_pred = X * w + b
    error = y_pred - y
    dw = 2 * np.mean(error * X)  # Gradient par rapport au poids
    db = 2 * np.mean(error)      # Gradient par rapport au biais

    # Mise √† jour des param√®tres
    w -= learning_rate * dw
    b -= learning_rate * db

    # Calcul de la perte
    loss = compute_loss(X, y, w, b)
    history.append(loss)

    # Mettre √† jour la ligne ajust√©e
    line.set_data(X, X * w + b)

    # Mettre √† jour la courbe de la perte
    loss_line.set_data(range(step + 1), history)

    return line, loss_line

# Cr√©er l'animation
ani = FuncAnimation(fig, update, frames=steps, blit=False, interval=100)

# Afficher l'animation
plt.tight_layout()
plt.show()  # Affiche l'animation
````

**format GIF**

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Param√®tres de la simulation
n = 100  # Nombre de particules
n_steps = 200  # Nombre de pas de temps
radius = 10  # Rayon de la bo√Æte

# Initialisation des positions et vitesses des particules
positions = np.random.uniform(-radius, radius, size=(n, 2))
velocities = np.random.uniform(-1, 1, size=(n, 2))

# Initialisation de la figure et des axes
fig, ax = plt.subplots()
ax.set_xlim(-radius, radius)
ax.set_ylim(-radius, radius)
scat = ax.scatter(positions[:, 0], positions[:, 1])

# Mise √† jour de la position des particules √† chaque √©tape
def update(frame):
    global positions, velocities

    # Mise √† jour des positions selon les vitesses
    positions += velocities

    # Collision avec les bords (r√©flexion)
    for i in range(n):
        if positions[i, 0] < -radius or positions[i, 0] > radius:
            velocities[i, 0] = -velocities[i, 0]
        if positions[i, 1] < -radius or positions[i, 1] > radius:
            velocities[i, 1] = -velocities[i, 1]

    # Mise √† jour des points dans la figure
    scat.set_offsets(positions)
    return scat,

# Cr√©ation de l'animation
ani = FuncAnimation(fig, update, frames=n_steps, interval=50, blit=True)

# Enregistrement de l'animation au format GIF
ani.save('animation.gif', writer='pillow')

# Affichage du GIF dans Jupyter
from IPython.display import Image
Image(filename='animation.gif')
````
