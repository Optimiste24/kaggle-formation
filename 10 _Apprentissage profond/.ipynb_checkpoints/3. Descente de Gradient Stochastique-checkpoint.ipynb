## Introduction  
Dans les deux premières leçons, nous avons appris à construire des réseaux entièrement connectés à partir de piles de couches denses. Lorsqu'il est créé, tous les poids du réseau sont initialisés de manière aléatoire — le réseau ne "**sait**" rien encore. Dans cette leçon, nous allons voir comment **entraîner un réseau de neurones**, c'est-à-dire comment les réseaux de neurones apprennent.

Comme pour toutes les tâches d'**apprentissage automatique**, nous commençons avec un ensemble de **données d'entraînement**. Chaque exemple dans les données d'entraînement est composé de certaines caractéristiques (**les entrées**) ainsi que d'une cible attendue (**la sortie**). Entraîner le réseau signifie ajuster ses poids de manière à ce qu'il puisse transformer les caractéristiques en la cible. 

Par exemple, dans le jeu de données des 80 céréales, nous voulons un réseau capable de prendre la teneur en '**sucre**', '**fibres**' et '**protéines**' de chaque céréale et de produire une prédiction pour les '**calories**' de cette céréale. Si nous parvenons à entraîner un réseau pour faire cela, ses poids doivent représenter d'une certaine manière la relation entre ces caractéristiques et cette cible, telle qu'elle est exprimée dans les données d'entraînement.

En plus des données d'entraînement, nous avons besoin de deux choses supplémentaires :

1. Une **fonction de perte** qui mesure la qualité des prédictions du réseau.
2. Un **optimiseur** qui peut indiquer au réseau comment modifier ses poids.

---
### **La Fonction de Perte**  
Nous avons vu comment concevoir une architecture pour un réseau, mais nous n'avons pas vu comment indiquer au réseau quel problème résoudre. C'est le rôle de la fonction de perte.

**La fonction de perte mesure l'écart entre la vraie valeur de la cible et la valeur prédite par le modèle.**

Différents problèmes nécessitent différentes fonctions de perte. Nous avons examiné des **problèmes de régression**, où la tâche consiste à prédire une valeur numérique — comme les calories dans les 80 céréales ou la note dans la qualité du vin rouge. D'autres tâches de régression pourraient être la prédiction du prix d'une maison ou de l'efficacité énergétique d'une voiture.

Une fonction de perte courante pour les problèmes de régression est **l'erreur absolue moyenne (MAE)**. Pour chaque prédiction `y_pred`, la **MAE** mesure l'écart par rapport à la vraie cible `y_true` par une différence absolue `abs(y_true - y_pred)`.

**La perte MAE totale sur un ensemble de données est la moyenne de toutes ces différences absolues.**

Un graphique représentant les barres d'erreur des points de données à la courbe ajustée.  
L'erreur absolue moyenne est la longueur moyenne entre la courbe ajustée et les points de données.

Outre la MAE, d'autres fonctions de perte que vous pourriez rencontrer pour les problèmes de régression sont l'**erreur quadratique moyenne (MSE)** ou la **perte de Huber** (toutes deux disponibles dans Keras).

Pendant l'entraînement, le modèle utilisera la fonction de perte comme guide pour trouver les valeurs correctes de ses poids (**une perte plus faible est meilleure**). En d'autres termes, la fonction de perte indique au réseau son objectif.


---
## **L'Optimiseur - Descente de Gradient Stochastique (SGD)**  
Nous avons décrit le problème que nous voulons que le réseau résolve, mais maintenant nous devons dire comment le résoudre. C'est le rôle de l'optimiseur. **L'optimiseur est un algorithme qui ajuste les poids pour minimiser la perte**.

Pratiquement tous les algorithmes d'optimisation utilisés en apprentissage profond appartiennent à une famille appelée **descente de gradient stochastique**. Ce sont des algorithmes **itératifs** qui entraînent un réseau par étapes. Une étape d'entraînement se déroule comme suit :

- 1. Échantillonner quelques données d'entraînement et les faire passer à travers le réseau pour faire des prédictions.
- 2. Mesurer la perte entre les prédictions et les vraies valeurs.
- 3. Enfin, ajuster les poids dans une direction qui réduit la perte.

Ensuite, répétez ce processus jusqu'à ce que la perte soit aussi petite que vous le souhaitez (ou jusqu'à ce qu'elle ne diminue plus).

Ajuster une ligne lot par lot. La perte diminue et les poids approchent leurs vraies valeurs.  
Entraîner un réseau de neurones avec la descente de gradient stochastique.

Chaque échantillon de données d'entraînement dans une **itération** est appelé un **minibatch** (ou souvent simplement "**batch**"), tandis qu'un cycle complet des données d'entraînement est appelé **une époque**. Le nombre d'époques pour lesquelles vous entraînez le réseau correspond au nombre de fois que le réseau verra chaque exemple d'entraînement.

L'animation montre le modèle linéaire de la leçon 1 entraîné avec SGD. Les points rouges pâles représentent l'ensemble de l'ensemble d'entraînement, tandis que les points rouges pleins sont les minibatchs. Chaque fois que SGD voit un nouveau minibatch, il déplace les poids (w la pente et b l'ordonnée à l'origine) vers leurs valeurs correctes pour ce lot. Lot après lot, la ligne finit par converger vers son meilleur ajustement. Vous pouvez voir que la perte diminue à mesure que les poids se rapprochent de leurs vraies valeurs.


---
## **Taux d'Apprentissage et Taille des Lots** 
Remarquez que la ligne ne fait qu'un petit décalage dans la direction de chaque lot (au lieu de se déplacer complètement). La taille de ces décalages est déterminée par le taux d'apprentissage. Un **taux d'apprentissage** plus petit signifie que le réseau doit voir plus de minibatchs avant que ses poids ne convergent vers leurs meilleures valeurs.

Le taux d'apprentissage et la taille des minibatchs sont les deux paramètres qui ont le plus grand effet sur la façon dont l'entraînement SGD se déroule. Leur interaction est souvent subtile et le bon choix pour ces paramètres n'est pas toujours évident. (Nous explorerons ces effets dans l'exercice.)

Heureusement, pour la plupart des travaux, il ne sera pas nécessaire de faire une recherche approfondie des hyperparamètres pour obtenir des résultats satisfaisants. 
**Adam** est un algorithme **SGD** qui a un taux d'apprentissage adaptatif, ce qui le rend adapté à la plupart des problèmes sans réglage des paramètres (il est en quelque sorte "**auto-ajustable**"). **Adam est un excellent optimiseur polyvalent**.


## **Ajouter la Fonction de Perte et l'Optimiseur**  
Après avoir défini un modèle, vous pouvez ajouter une fonction de perte et un optimiseur avec la méthode `compile` du modèle :


```
model.compile(
    optimizer="adam",
    loss="mae",
)
```

Remarquez que nous pouvons spécifier la perte et l'optimiseur simplement avec une chaîne de caractères. Vous pouvez également y accéder directement via l'**API Keras** — si vous souhaitez ajuster des paramètres, par exemple — mais pour nous, les valeurs par défaut fonctionneront bien.

## **Que Signifie le Nom ?** 
Le **gradient** est un vecteur qui nous indique dans quelle direction les poids doivent aller. Plus précisément, il nous indique comment modifier les poids pour que la perte change le plus rapidement. Nous appelons notre processus "**descente de gradient**" car il utilise le gradient pour descendre la courbe de perte vers un minimum. "**Stochastique**" signifie "déterminé par le hasard." Notre entraînement est stochastique car les minibatchs sont des échantillons aléatoires de l'ensemble de données. Et c'est pourquoi on l'appelle **SGD** !

# **Exemple - Qualité du Vin Rouge**  
Maintenant, nous savons tout ce dont nous avons besoin pour commencer à entraîner des modèles d'apprentissage profond. Voyons cela en action ! Nous allons utiliser le jeu de données sur la qualité du vin rouge.

Ce jeu de données contient des mesures physico-chimiques d'environ 1600 vins rouges portugais. Il inclut également une note de qualité pour chaque vin, issue de tests de dégustation à l'aveugle. **À quel point pouvons-nous prédire la qualité perçue d'un vin à partir de ces mesures ?**

Nous avons intégré toute la préparation des données dans la cellule suivante. Ce n'est pas essentiel pour ce qui suit, alors n'hésitez pas à la sauter. Une chose que vous pourriez noter pour l'instant est que nous avons redimensionné chaque caractéristique pour qu'elle se situe dans l'intervalle [0, 1]. Comme nous en discuterons davantage dans la leçon 5, les réseaux de neurones ont tendance à mieux performer lorsque leurs entrées sont sur une échelle commune.


````python
import pandas as pd
from IPython.display import display

red_wine = pd.read_csv('red-wine.csv')

# Créer des ensembles d'entraînement et de validation
df_train = red_wine.sample(frac=0.7, random_state=0)
df_valid = red_wine.drop(df_train.index)
display(df_train.head(4))

# Redimensionner à [0, 1]
max_ = df_train.max(axis=0)
min_ = df_train.min(axis=0)
df_train = (df_train - min_) / (max_ - min_)
df_valid = (df_valid - min_) / (max_ - min_)

# Séparer les caractéristiques et la cible
X_train = df_train.drop('quality', axis=1)
X_valid = df_valid.drop('quality', axis=1)
y_train = df_train['quality']
y_valid = df_valid['quality']
````

**Combien d'entrées ce réseau devrait-il avoir ?** Nous pouvons le découvrir en regardant le nombre de colonnes dans la matrice de données. Assurez-vous de ne pas inclure la cible ('quality') ici — uniquement les caractéristiques d'entrée.


````python
print(X_train.shape)
````

Onze colonnes signifient onze entrées.

Nous avons choisi un réseau à trois couches avec plus de 1500 neurones. Ce réseau devrait être capable d'apprendre des relations assez complexes dans les données.


````python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Input(shape=[11]), 
    layers.Dense(512, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(1),
])
````

Déterminer l'**architecture** de votre modèle devrait faire partie d'un processus. Commencez simplement et utilisez la **perte de validation** comme guide. Vous en apprendrez plus sur le développement de modèles dans les exercices.

Après avoir défini le modèle, nous **compilons l'optimiseur et la fonction de perte**.

````python
model.compile(
    optimizer='adam',
    loss='mae',
)
````

Maintenant, nous sommes prêts à commencer l'entraînement ! Nous avons dit à **Keras** de donner à l'optimiseur **256 lignes** des données d'entraînement à la fois (le `batch_size`) et de le faire **10 fois** à travers l'ensemble de données (les `epochs`).


````python
history = model.fit(
    X_train, y_train,
    validation_data=(X_valid, y_valid),
    batch_size=256,
    epochs=10,
)
````

Vous pouvez voir que Keras vous tiendra informé de la perte pendant que le modèle s'entraîne.

Souvent, une meilleure façon de visualiser la perte est de la tracer. La méthode `fit` conserve en fait un enregistrement de la perte produite pendant l'entraînement dans un objet `History`. Nous allons convertir les données en un dataframe Pandas, ce qui facilite le traçage.


````python
import pandas as pd

# Convertir l'historique d'entraînement en dataframe
history_df = pd.DataFrame(history.history)
# Utiliser la méthode de traçage native de Pandas
history_df['loss'].plot();
````

Remarquez comment la perte se stabilise à mesure que les époques passent. Lorsque la courbe de perte devient horizontale comme cela, cela signifie que le modèle a appris tout ce qu'il pouvait et qu'il n'y a aucune raison de continuer pour des époques supplémentaires.

### **Présentation du dataset Fuel Economy 🚗**  
Dans ce dataset, votre objectif est de **prédire la consommation de carburant d'un véhicule** (*Fuel Economy - FE*) à partir de plusieurs caractéristiques, comme **le type de moteur, l'année de fabrication**, etc.  

📌 **Chargement des données :** 

````python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer, make_column_selector
from sklearn.model_selection import train_test_split

# Chargement des données
fuel = pd.read_csv('fuel.csv')

# Copie des données
X = fuel.copy()
y = X.pop('FE')  # Suppression de la colonne cible (FE) des caractéristiques

X.head()
````

📌 **Prétraitement des données :** 

````python
preprocessor = make_column_transformer(
    (StandardScaler(), make_column_selector(dtype_include=np.number)),  # Normalisation des variables numériques
    (OneHotEncoder(sparse_output=False), make_column_selector(dtype_include=object)),  # Encodage des variables catégorielles
)

X = preprocessor.fit_transform(X)
y = np.log(y)  # Transformation logarithmique de la cible (au lieu de la standardiser)

input_shape = [X.shape[1]]  # Taille des entrées
print("Input shape: {}".format(input_shape))  # Affichage de la dimension des entrées
````

### **Définition du modèle de réseau de neurones 🧠**  
On définit un **réseau de neurones** comportant **trois couches cachées** et une couche de sortie :  
- **2 couches cachées** de **128 neurones** avec activation **ReLU**  
- **1 couche cachée** de **64 neurones** avec activation **ReLU**  
- **1 couche de sortie** avec **1 neurone** (sans activation, car il s'agit d'un problème de régression)  

---

### 🧐 **Résumé**  
| Couche | Nombre de neurones | Activation | Raison |
|--------|------------------|------------|--------|
| **Cachée 1** | 128 | ReLU | Capturer des relations complexes |
| **Cachée 2** | 128 | ReLU | Approfondir l'apprentissage |
| **Cachée 3** | 64  | ReLU | Réduction progressive des dimensions |
| **Sortie** | 1 | Aucune | Prédire un nombre réel (régression) |

### **💡 Pourquoi pas plus ou moins de couches ?**  
- **Plus de couches** : Peut améliorer la performance, mais augmente le risque de **sur-apprentissage** et rend l'entraînement plus lent.  
- **Moins de couches** : Peut ne pas capturer suffisamment les **relations complexes** entre les variables.  

👉 **En pratique, ces hyperparamètres (nombre de couches, neurones, activation) sont souvent optimisés par validation croisée et ajustements successifs.** 🔧

````python
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Input(shape=input_shape),  # Définition explicite de l'entrée
    layers.Dense(128, activation='relu'),  # Couche cachée 1
    layers.Dense(128, activation='relu'),  # Couche cachée 2
    layers.Dense(64, activation='relu'),  # Couche cachée 3
    layers.Dense(1),  # Couche de sortie
])
````

Ce modèle est prêt à être entraîné pour prédire la consommation de carburant des voitures ! 🚗💨

## Question 1) Ajouter la fonction de perte et l'optimiseur  
Avant d'entraîner le réseau, nous devons définir **la fonction de perte (loss function) et l'optimiseur (optimizer)** que nous allons utiliser.  

En utilisant la méthode `.compile()` du modèle, ajoutez :  
- **L'optimiseur Adam** (*Adaptive Moment Estimation*), qui est un optimiseur avancé combinant les avantages de SGD et RMSprop.  
- **La fonction de perte MAE** (*Mean Absolute Error*), qui est couramment utilisée pour les problèmes de régression car elle mesure l'écart moyen absolu entre les valeurs prédites et réelles.  

````python
# Définition de la fonction de perte et de l'optimiseur
model.compile(optimizer="adam", loss="mae")
````

## Question 2) Entraîner le modèle

##### Pourquoi **200** et pas une autre valeur ?
1. **Assez d'itérations pour l'apprentissage**

   - Après plusieurs époques, l'amélioration devient plus lente.

2. **Expérience et empirisme**  
   - Pour des problèmes similaires en **régression avec des réseaux de neurones**, 100 à 200 époques sont souvent suffisantes.  
   - On peut ajuster ce nombre après observation des performances.  

3. **Observation du comportement du modèle**  
   - Si la **perte diminue toujours** après 200 époques, il faudra peut-être augmenter le nombre d'époques.  
   - Si la **perte se stabilise bien avant**, on pourrait réduire le nombre d'époques pour **gagner du temps de calcul**.

La question 3 porte sur l'évaluation de l'entraînement du modèle et sur la manière dont la perte évolue au cours du temps d'entraînement. Voici une explication détaillée pour mieux comprendre la réponse.

### Évaluation de l'entraînement

L'idée centrale ici est de se demander si la perte continuera de diminuer si vous entraînez le modèle plus longtemps. En d'autres termes, si vous ajoutez plus d'époques d'entraînement, est-ce que la perte (la fonction de coût) continuera à baisser ou s'atteindra-t-elle un minimum ? Cela dépend de l'évolution de la courbe de la perte pendant l'entraînement.

#### Scénarios :

1. **Si la courbe d'apprentissage s'est stabilisée** :
   - Si la perte a atteint un point où elle reste constante ou fluctue peu, alors **il n'y a généralement aucun avantage** à entraîner le modèle plus longtemps. Cela signifie que le modèle a déjà convergé vers un minimum local et qu'il est peu probable que l'ajout de nouvelles époques améliore les résultats.
   
2. **Si la perte continue de diminuer** :
   - Si la perte continue de baisser, cela signifie que le modèle est encore en train d'apprendre et pourrait bénéficier de plus d'époques d'entraînement. Dans ce cas, l'entraînement prolongé pourrait effectivement **améliorer les performances** du modèle.

### L'impact du taux d'apprentissage (learning rate) et de la taille des lots (batch size)

Le **taux d'apprentissage** et la **taille des lots** influencent plusieurs aspects de l'entraînement :

- **Le temps nécessaire pour entraîner le modèle** : Un grand taux d'apprentissage peut rendre l'entraînement plus rapide, mais cela peut aussi rendre le processus plus instable.
  
- **Le bruit des courbes d'apprentissage** : Des petites tailles de lots augmentent le bruit (variabilité) des courbes d'apprentissage, car chaque lot ne représente qu'une petite fraction des données. À l'inverse, de grands lots donnent des mises à jour plus stables, mais peuvent rendre l'entraînement plus lent.

- **La valeur minimale que peut atteindre la perte** : Un taux d'apprentissage trop élevé peut entraîner une perte de qualité dans l'entraînement, empêchant le modèle de converger vers un bon minimum.

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Fonction de descente de gradient stochastique
def sgd_step(w, b, x, y, learning_rate):
    # Calculer les gradients
    gradient_w = 2 * np.mean((w * x + b - y) * x)
    gradient_b = 2 * np.mean(w * x + b - y)
    
    # Mettre à jour les poids
    w -= learning_rate * gradient_w
    b -= learning_rate * gradient_b
    return w, b

# Fonction d'animation de la descente de gradient stochastique
def animate_sgd(learning_rate, batch_size, num_examples, steps, true_w, true_b):
    # Créer des données simulées
    X = np.random.rand(num_examples)
    Y = true_w * X + true_b + np.random.normal(scale=0.1, size=num_examples)

    # Initialiser les poids
    w, b = np.random.randn(), np.random.randn()

    fig, ax = plt.subplots()
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)

    # Line et point d'animation
    line, = ax.plot([], [], lw=2)
    point, = ax.plot([], [], 'ro')

    # Fonction d'initialisation
    def init():
        line.set_data([], [])
        point.set_data([], [])
        return line, point

    # Fonction de mise à jour
    def update(frame):
        nonlocal w, b
        # Effectuer une étape de descente de gradient
        w, b = sgd_step(w, b, X[:batch_size], Y[:batch_size], learning_rate)
        # Mettre à jour la ligne et les points
        line.set_data(X, w * X + b)
        point.set_data(w, b)
        return line, point

    # Créer l'animation
    ani = FuncAnimation(fig, update, frames=steps, init_func=init, blit=True)
    return ani

# Liste des différentes combinaisons à tester
param_combinations = [
    (0.05, 32, 256),
    (0.05, 2, 256),
    (0.05, 128, 256),
    (0.02, 32, 256),
    (0.2, 32, 256),
    (1.0, 32, 256),
    (0.9, 4096, 8192),
    (0.99, 4096, 8192)
]

# Tester chaque combinaison
for learning_rate, batch_size, num_examples in param_combinations:
    print(f"Testing with: Learning Rate = {learning_rate}, Batch Size = {batch_size}, Num Examples = {num_examples}")
    
    # Exécution de l'animation avec les paramètres actuels
    ani = animate_sgd(
        learning_rate=learning_rate,
        batch_size=batch_size,
        num_examples=num_examples,
        steps=50,  # total training steps (batches seen)
        true_w=3.0,  # the slope of the data
        true_b=2.0,  # the bias of the data
    )
    
    # Forcer l'affichage de l'animation
    plt.show()
````

## **Créer la fonction animate_sgd**

````python
import numpy as np
import matplotlib.pyplot as plt

def animate_sgd(learning_rate=0.05, batch_size=32, num_examples=256, steps=50, true_w=3.0, true_b=2.0):
    # Initialisation des paramètres
    w = np.random.randn()  # Poids initial
    b = np.random.randn()  # Biais initial
    X = np.random.randn(num_examples, 1)  # Données d'entrée
    y = X * true_w + true_b + np.random.randn(num_examples, 1)  # Données de sortie (avec bruit)

    # Variables pour stocker l'évolution du modèle
    ws, bs = [], []

    for step in range(steps):
        # Calcul des gradients (descente de gradient)
        y_pred = X * w + b
        error = y_pred - y
        dw = 2 * np.mean(error * X)  # Gradient par rapport au poids
        db = 2 * np.mean(error)      # Gradient par rapport au biais

        # Mise à jour des paramètres
        w -= learning_rate * dw
        b -= learning_rate * db

        # Sauvegarder l'évolution pour l'animation
        ws.append(w)
        bs.append(b)

    # Affichage de l'animation
    plt.plot(ws, label="Poids")
    plt.plot(bs, label="Biais")
    plt.legend()
    plt.xlabel('Étape')
    plt.ylabel('Valeur')
    plt.title('Animation de la descente de gradient')
    plt.show()

# Appel de la fonction
animate_sgd()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Générer des données d'exemple
true_w = 3.0  # Pente réelle
true_b = 2.0  # Biais réel
num_examples = 256
X = np.random.randn(num_examples, 1)  # Données d'entrée
y = X * true_w + true_b + np.random.randn(num_examples, 1)  # Données de sortie avec bruit

# Initialisation des paramètres du modèle
w = np.random.randn()
b = np.random.randn()

# Paramètres de l'entraînement
learning_rate = 0.1
steps = 100
history = []  # Pour stocker l'évolution de la perte

# Fonction de calcul de la perte (MSE)
def compute_loss(X, y, w, b):
    y_pred = X * w + b
    return np.mean((y_pred - y) ** 2)

# Préparation de l'animation
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.set_title('Fitted Line')
ax1.set_xlabel('X')
ax1.set_ylabel('y')
ax2.set_title('Training Loss')
ax2.set_xlabel('Steps')
ax2.set_ylabel('Loss')

line, = ax1.plot([], [], label='Line ajustée', color='red')
loss_line, = ax2.plot([], [], label='Training Loss', color='blue')
ax1.scatter(X, y, alpha=0.5)

# Fonction de mise à jour de l'animation
def update(step):
    global w, b
    # Calcul des prédictions et de l'erreur
    y_pred = X * w + b
    error = y_pred - y
    dw = 2 * np.mean(error * X)  # Gradient par rapport au poids
    db = 2 * np.mean(error)      # Gradient par rapport au biais

    # Mise à jour des paramètres
    w -= learning_rate * dw
    b -= learning_rate * db

    # Calcul de la perte
    loss = compute_loss(X, y, w, b)
    history.append(loss)

    # Mettre à jour la ligne ajustée
    line.set_data(X, X * w + b)

    # Mettre à jour la courbe de la perte
    loss_line.set_data(range(step + 1), history)

    return line, loss_line

# Créer l'animation
ani = FuncAnimation(fig, update, frames=steps, blit=False, interval=100)

# Afficher l'animation
plt.tight_layout()
plt.show()  # Affiche l'animation
````

**format GIF**

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Paramètres de la simulation
n = 100  # Nombre de particules
n_steps = 200  # Nombre de pas de temps
radius = 10  # Rayon de la boîte

# Initialisation des positions et vitesses des particules
positions = np.random.uniform(-radius, radius, size=(n, 2))
velocities = np.random.uniform(-1, 1, size=(n, 2))

# Initialisation de la figure et des axes
fig, ax = plt.subplots()
ax.set_xlim(-radius, radius)
ax.set_ylim(-radius, radius)
scat = ax.scatter(positions[:, 0], positions[:, 1])

# Mise à jour de la position des particules à chaque étape
def update(frame):
    global positions, velocities

    # Mise à jour des positions selon les vitesses
    positions += velocities

    # Collision avec les bords (réflexion)
    for i in range(n):
        if positions[i, 0] < -radius or positions[i, 0] > radius:
            velocities[i, 0] = -velocities[i, 0]
        if positions[i, 1] < -radius or positions[i, 1] > radius:
            velocities[i, 1] = -velocities[i, 1]

    # Mise à jour des points dans la figure
    scat.set_offsets(positions)
    return scat,

# Création de l'animation
ani = FuncAnimation(fig, update, frames=n_steps, interval=50, blit=True)

# Enregistrement de l'animation au format GIF
ani.save('animation.gif', writer='pillow')

# Affichage du GIF dans Jupyter
from IPython.display import Image
Image(filename='animation.gif')
````

## Question 4) Taux d'apprentissage et taille de lot 
Quel effet ont eu ces changements de paramètres ? Après y avoir réfléchi, exécutez la cellule ci-dessous pour une discussion.



**Correct :**

Vous avez probablement remarqué que des tailles de lots plus petites entraînaient des mises à jour de poids et des courbes de perte plus bruyantes. Cela est dû au fait que chaque lot est un petit échantillon de données et que les petits échantillons ont tendance à donner des estimations plus bruyantes. Cependant, les petits lots peuvent avoir un effet de « moyennage », ce qui peut être bénéfique.

Des taux d'apprentissage plus petits rendent les mises à jour plus petites et l'entraînement prend plus de temps pour converger. De grands taux d'apprentissage peuvent accélérer l'entraînement, mais n'atteignent pas aussi bien un minimum. Lorsque le taux d'apprentissage est trop grand, l'entraînement peut échouer complètement. (Essayez de régler le taux d'apprentissage sur une valeur élevée comme 0,99 pour observer cela.)
