# **As & With**
Organisez votre requête pour une meilleure lisibilité. Cela devient particulièrement important pour les requêtes complexes.

### Introduction
Avec tout ce que vous avez appris, vos requêtes SQL deviennent assez longues, ce qui peut les rendre difficiles à comprendre (et à déboguer).

Vous allez maintenant apprendre à utiliser **AS et WITH** pour organiser vos requêtes et les rendre plus faciles à lire.

En chemin, nous utiliserons la table familière **pets**, mais cette fois-ci, elle inclut l'âge des animaux.

---

## AS
Vous avez appris dans un tutoriel précédent comment utiliser **AS** pour renommer les colonnes générées par vos requêtes, ce qu'on appelle aussi **l'aliasage**. C'est similaire à la façon dont Python utilise **as** pour l'aliasage lors d'importations comme **import pandas as pd** ou **import seaborn as sns**.

Pour utiliser **AS** en SQL, insérez-le juste après la colonne que vous sélectionnez. Voici un exemple de requête sans clause AS :
```
SELECT COUNT(ID)
FROM `pets`;
```
- Et voici un exemple de la même requête, mais avec AS :
```
SELECT COUNT(ID) AS Number
FROM `pets`;
```
Ces requêtes renvoient les mêmes informations, mais dans la deuxième requête, la colonne renvoyée par la fonction COUNT() sera appelée Number, plutôt que le nom par défaut f0__.

---
## WITH ... AS
À lui seul, AS est un moyen pratique de nettoyer les données renvoyées par votre requête. Il est encore plus puissant lorsqu'il est combiné avec **WITH** dans ce qu'on appelle une **expression de table commune** (ou **CTE, pour Common Table Expression**).

Une CTE est *une table temporaire* que vous créez dans votre requête. Les CTE sont utiles pour diviser vos requêtes en morceaux lisibles, et vous pouvez écrire des requêtes sur ces tables temporaires.

Par exemple, vous pourriez vouloir utiliser la table **pets** pour poser des questions sur les animaux plus âgés en particulier. 
- Vous pouvez commencer par créer une CTE qui ne contient que des informations sur les animaux de plus de cinq ans, comme ceci :
```
WITH Seniors AS
(
    SELECT *
    FROM `pets`
    WHERE Age > 5
)
```
Bien que cette requête incomplète ne renvoie rien, elle crée une CTE que nous pouvons ensuite référencer (comme Seniors) pour écrire le reste de la requête.

Nous pouvons terminer la requête en extrayant les informations que nous voulons de la CTE. 
- La requête complète ci-dessous crée d'abord la CTE, puis renvoie tous les IDs qu'elle contient.
```
WITH Seniors AS
(
    SELECT *
    FROM `pets`
    WHERE Age > 5
)
SELECT ID
FROM Seniors;
```
Vous pourriez faire cela sans CTE, mais si c'était la première partie d'une très longue requête, supprimer la CTE rendrait la requête beaucoup plus difficile à suivre.

Il est également important de noter que les *CTE n'existent que dans la requête où vous les créez*, et vous ne pouvez pas les référencer dans des requêtes ultérieures. Ainsi, toute requête utilisant une CTE est toujours divisée en deux parties :
- D'abord, nous créons la CTE,
- Ensuite, nous écrivons une requête qui utilise la CTE.

---
## Exemple : Combien de transactions Bitcoin sont effectuées par mois ?
Nous allons utiliser une CTE pour déterminer combien de transactions Bitcoin ont été effectuées chaque jour sur toute la durée d'un ensemble de données de transactions Bitcoin.

Nous allons examiner la table **transactions**. Voici un aperçu des premières lignes. (Le code correspondant est masqué, mais vous pouvez le démasquer en cliquant sur le bouton "Code" ci-dessous.)

````python
from google.cloud import bigquery

# Créer un objet "Client"
client = bigquery.Client()

# Construire une référence au dataset "crypto_bitcoin"
dataset_ref = client.dataset("crypto_bitcoin", project="bigquery-public-data")

# Requête API - récupérer le dataset
dataset = client.get_dataset(dataset_ref)

# Construire une référence à la table "transactions"
table_ref = dataset_ref.table("transactions")

# Requête API - récupérer la table
table = client.get_table(table_ref)

# Prévisualiser les cinq premières lignes de la table "transactions"
client.list_rows(table, max_results=5).to_dataframe()
````

Puisque la colonne **block_timestamp** contient la date de chaque transaction au format **DATETIME**, nous allons convertir ces dates en format **DATE** en utilisant la commande DATE().

Nous faisons cela en utilisant une CTE, puis la partie suivante de la requête compte le nombre de transactions pour chaque date et trie la table pour que les dates les plus anciennes apparaissent en premier.

````python
# Requête pour sélectionner le nombre de transactions par date, triées par date
query_with_CTE = """ 
                 WITH time AS 
                 (
                     SELECT DATE(block_timestamp) AS trans_date
                     FROM `bigquery-public-data.crypto_bitcoin.transactions`
                 )
                 SELECT COUNT(1) AS transactions,
                        trans_date
                 FROM time
                 GROUP BY trans_date
                 ORDER BY trans_date
                 """

# Configurer la requête (annuler la requête si elle utilise trop de quota, avec une limite fixée à 10 Go)
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=10**10)
query_job = client.query(query_with_CTE, job_config=safe_config)

# Requête API - exécuter la requête et convertir les résultats en DataFrame pandas
transactions_by_date = query_job.to_dataframe()

# Afficher les cinq premières lignes
transactions_by_date.head()
````

- Comme les résultats sont triés, nous pouvons facilement tracer les résultats bruts pour montrer le nombre de transactions Bitcoin par jour sur toute la durée de cet ensemble de données.

````python
transactions_by_date.set_index('trans_date').plot()
````

**Comme vous pouvez le voir, les expressions de table communes (CTE) vous permettent de déplacer une grande partie de votre nettoyage de données dans SQL. C'est particulièrement utile dans le cas de BigQuery, car c'est beaucoup plus rapide que de faire ce travail dans Pandas.**

# Question 3: Déterminer la période des données
Si les données sont trop anciennes, il faut être prudent avant de supposer qu'elles sont toujours représentatives des tendances actuelles du trafic. **Écrivez une requête qui compte le nombre de trajets par année.**

Vos résultats doivent comporter deux colonnes :
- **year** : l'année des trajets
- **num_trips** : le nombre de trajets pour cette année

**Indices :**
- Lorsque vous utilisez **GROUP BY et ORDER BY**, vous devez vous référer aux colonnes par l'alias **year** que vous avez défini en haut de la requête **SELECT**.
- Le code SQL pour extraire l'année à partir de **trip_start_timestamp** est :
```
SELECT EXTRACT(YEAR FROM trip_start_timestamp)
```
- Le format du champ **FROM** peut être un peu délicat au début. Il suit cette structure :
1. Un accent grave (`)
2. Le nom du projet : ici, **bigquery-public-data**
3. Un point (.)
4. Le nom de l'ensemble de données : ici, **chicago_taxi_trips**
5. Un point (.)
6. Le nom de la table (réponse trouvée à l'étape 1)
7. Un accent grave (`)

````python
# Votre requête SQL pour compter les trajets par année
rides_per_year_query = """
SELECT 
    EXTRACT(YEAR FROM trip_start_timestamp) AS year, 
    COUNT(*) AS num_trips
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
GROUP BY year
ORDER BY year
"""

# Configuration pour limiter l'utilisation du quota
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=10**10)

# Exécution de la requête
rides_per_year_query_job = client.query(rides_per_year_query, job_config=safe_config)

# Conversion des résultats en DataFrame pandas
rides_per_year_result = rides_per_year_query_job.to_dataframe()

# Affichage des résultats
print(rides_per_year_result)
````

Les données s'étendent de **2013 à 2023**, avec une baisse significative du nombre de trajets après 2019.  

### Observations :  
- **2014** a le plus grand nombre de trajets (~37,4M), suivi de **2013** et **2015**.  
- À partir de **2017**, on observe une diminution progressive.  
- **2020** (début de la pandémie COVID-19) montre une chute drastique (~3,9M trajets, soit environ **25% du volume de 2019**).  
- Une légère reprise est visible entre **2021 et 2023**, mais sans retrouver les niveaux d'avant 2020.  

### Conséquences pour l'analyse :  
- Les tendances actuelles du trafic peuvent être très différentes de celles observées avant 2020.  
- Si nous analysons l'impact du volume du trafic sur la vitesse des trajets, il faut considérer ces variations.  
- Une séparation entre **périodes pré- et post-COVID** pourrait être pertinente.

## Question 4: Approfondir l'analyse
Vous souhaitez examiner de plus près les trajets de 2016. Copiez la requête SQL utilisée précédemment dans rides_per_year_query et adaptez-la dans rides_per_month_query en effectuant deux modifications :
- Ajoutez une clause **WHERE** pour ne conserver que les données de **2016**.
- Modifiez l'extraction de l'année (**YEAR**) vers le mois (**MONTH**).

````python

rides_per_month_query = """
SELECT 
    EXTRACT(MONTH FROM trip_start_timestamp) AS month, 
    COUNT(*) AS num_trips 
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
WHERE EXTRACT(YEAR FROM trip_start_timestamp) = 2016
GROUP BY month
ORDER BY month
"""

# Configuration de la requête pour éviter les dépassements de quota
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=10**10)
rides_per_month_query_job = client.query(rides_per_month_query, job_config=safe_config)

# Exécution de la requête et conversion en DataFrame pandas
rides_per_month_result = rides_per_month_query_job.to_dataframe()

# Affichage des résultats
print(rides_per_month_result)
````

Les résultats montrent une **variation du nombre de trajets par mois en 2016**. Voici quelques observations intéressantes :  

📈 **Mars, avril et mai** ont enregistré le plus grand nombre de trajets, avec un pic en **mai (2 859 147 trajets)**.  
📉 **Novembre et décembre** ont les chiffres les plus bas, ce qui peut être dû au **froid hivernal** et aux **jours fériés** affectant les déplacements.  
🌞 **Juillet et août** affichent une légère baisse, peut-être en raison des **vacances estivales**.

## Question 5: Rédigez la requête
Il est temps d'augmenter la complexité de vos requêtes. Rédigez une requête qui affiche, pour **chaque heure de la journée**, le **nombre total de trajets** ainsi que la **vitesse moyenne** des courses.  

Votre résultat doit contenir **trois colonnes** :  

1️⃣ **hour_of_day** - L'heure d'extraction à partir de **trip_start_timestamp**, triée dans l'ordre croissant.  
2️⃣ **num_trips** - Le **nombre total de trajets** commencés à cette heure-là (exemple : combien de trajets ont débuté entre **6h et 7h du matin**, quel que soit le jour).  
3️⃣ **avg_mph** - La **vitesse moyenne en miles par heure** des trajets débutés à cette heure. Elle est calculée comme suit :  
   \[
   \text{avg_mph} = \frac{3600 \times \sum \text{(trip_miles)}}{\sum \text{(trip_seconds)}}
   \]
   (Le facteur **3600** convertit les secondes en heures).  

**Restrictions à appliquer** à la requête :  

✅ **Filtrer les courses entre le 1er janvier 2016 et le 1er avril 2016** (trip_start_timestamp entre ces deux dates).  
✅ **Exclure les trajets de durée ou distance nulle** (trip_seconds > 0 et trip_miles > 0).  

Vous utiliserez une **Common Table Expression (CTE)** pour pré-sélectionner uniquement les trajets pertinents. Comme cet ensemble de données est volumineux, cette CTE doit **ne sélectionner que les colonnes nécessaires** pour l'analyse finale.  

⚠️ **C'est une requête plus difficile que celles vues jusqu'à présent. Bonne chance !** 🚀  


````python

speeds_query = """
               WITH RelevantRides AS
               (
                   SELECT
                       EXTRACT(HOUR FROM trip_start_timestamp) AS hour_of_day,
                       trip_miles,
                       trip_seconds
                   FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
                   WHERE trip_start_timestamp > '2016-01-01' 
                         AND trip_start_timestamp < '2016-04-01'
                         AND trip_seconds > 0
                         AND trip_miles > 0
               )
               SELECT 
                   hour_of_day,
                   COUNT(*) AS num_trips,
                   (3600 * SUM(trip_miles) / SUM(trip_seconds)) AS avg_mph
               FROM RelevantRides
               GROUP BY hour_of_day
               ORDER BY hour_of_day
               """

# Configuration de la requête
safe_config = bigquery.QueryJobConfig(maximum_bytes_billed=10**10)
speeds_query_job = client.query(speeds_query, job_config=safe_config) # Lancer la requête

# Requête API - Exécution et récupération du DataFrame pandas
speeds_result = speeds_query_job.to_dataframe()

# Voir les résultats
print(speeds_result)
````

Le résultat montre le nombre de trajets (`num_trips`) et la vitesse moyenne (`avg_mph`) pour chaque heure de la journée entre janvier et avril 2016. Voici quelques observations intéressantes à partir des données :

- La vitesse moyenne semble assez variable tout au long de la journée, avec des pics tôt le matin (par exemple, à 5h, la vitesse moyenne est de 33.07 mph) et des baisses dans l'après-midi (par exemple, à 17h, la vitesse moyenne est de 14.64 mph).
- Les heures avec le plus grand nombre de trajets sont autour de 17h (pointe avec 394,324 trajets), ce qui est typique des périodes de forte circulation en fin de journée.

Cela confirme que les trajets effectués à différentes heures de la journée peuvent être influencés par des facteurs comme le trafic, ce qui se reflète dans les vitesses moyennes et les volumes de trajets. 😊
