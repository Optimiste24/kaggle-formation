### Transformations mathématiques
Les relations entre les fonctionnalités numériques sont souvent exprimées par des formules mathématiques que vous pouvez découvrir grâce à vos recherches. Avec Pandas, vous pouvez appliquer des opérations arithmétiques directement sur les colonnes, comme si elles étaient des nombres ordinaires.

**Exemple** :

Dans l'ensemble de données sur les automobiles, certaines fonctionnalités décrivent le moteur d'une voiture. La recherche propose des formules pour créer des fonctionnalités utiles. Par exemple, le "ratio de course" mesure l'efficacité d'un moteur par rapport à ses performances :

```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")

autos = pd.read_csv("autos.csv")
concrete = pd.read_csv("concrete.csv")
accidents = pd.read_csv("accidents.csv")
```

```python
autos.head()
```

```python
print(autos.columns)
```

```python
autos["stroke_ratio"] = autos.stroke / autos.bore
autos[["stroke", "bore", "stroke_ratio"]].head()
```

Pour des formules plus complexes, comme la "**cylindrée**" d'un moteur (une mesure de sa puissance) :

```python
autos["displacement"] = (
    np.pi * ((0.5 * autos.bore) ** 2) * autos.stroke * autos.num_of_cylinders
)
```

### Normalisation des données
La visualisation des données peut suggérer des transformations pour normaliser une fonctionnalité. Par exemple, la distribution de la vitesse du vent (**WindSpeed**) dans les accidents de la route aux États-Unis est fortement asymétrique. Ici, un logarithme est efficace :

```python
print(accidents.columns)
```

```python
# Si la variable présente des valeurs égales à 0.0, utilisez np.log1p (log(1+x)) au lieu de np.log
accidents["LogWindSpeed"] = accidents.WindSpeed.apply(np.log1p)

# Tracer une comparaison
fig, axs = plt.subplots(1, 2, figsize=(8, 4))
sns.kdeplot(accidents.WindSpeed, fill=True, ax=axs[0])
sns.kdeplot(accidents.LogWindSpeed, fill=True, ax=axs[1])
plt.show()
```

## Comptages
Les caractéristiques décrivant la présence ou l'absence de quelque chose sont souvent regroupées en ensembles, comme un ensemble de facteurs de risque pour une maladie, par exemple. Vous pouvez agréger ces caractéristiques en créant un comptage.

Ces caractéristiques seront binaires (1 pour Présent, 0 pour Absent) ou booléennes (Vrai ou Faux). En Python, les valeurs booléennes peuvent être additionnées de la même manière que des entiers.

Dans les données sur les accidents de la route, plusieurs caractéristiques indiquent si un objet routier était présent à proximité de l'accident. On peut créer un comptage total du nombre de ces caractéristiques routières à l'aide de la méthode sum :

```python
roadway_features = ["Amenity", "Bump", "Crossing", "GiveWay",
    "Junction", "NoExit", "Railway", "Roundabout", "Station", "Stop",
    "TrafficCalming", "TrafficSignal"]

accidents["RoadwayFeatures"] = accidents[roadway_features].sum(axis=1)

# Afficher les premières lignes des données
accidents[roadway_features + ["RoadwayFeatures"]].head(10)
```

---

#### Méthode intégrée des DataFrames pour créer des valeurs booléennes
Vous pouvez également utiliser les méthodes intégrées des DataFrames pour créer des valeurs booléennes. Dans le dataset **Concrete**, on trouve les quantités des composants utilisés dans une formulation de béton. Beaucoup de formulations manquent un ou plusieurs composants (**c'est-à-dire que la valeur d'un composant est 0**). Voici comment compter le nombre de composants présents dans une formulation en utilisant la méthode `gt` (supérieur à) des DataFrames :

```python
components = [ "Cement", "BlastFurnaceSlag", "FlyAsh", "Water",
               "Superplasticizer", "CoarseAggregate", "FineAggregate"]
concrete["Composants"] = concrete[components].gt(0).sum(axis=1)

# Afficher les premières lignes des données
concrete[components + ["Composants"]].head(10)
```

### Décomposer et combiner des caractéristiques
Il arrive souvent que des chaînes complexes puissent être décomposées en éléments plus simples. Voici quelques exemples courants :

- Numéros d'identification : `123-45-6789`
- Numéros de téléphone : `(999) 555-0123`
- Adresses : `8241 Kaggle Ln., Goose City, NV`
- Adresses Internet : `http://www.kaggle.com`
- Codes produits : `0 36000 29145 2`
- Dates et heures : `Lun 30 Sep 07:06:05 2013`

Ces caractéristiques ont souvent une structure que vous pouvez exploiter. Par exemple, les numéros de téléphone américains contiennent un indicatif régional (la partie (999)) qui indique la localisation de l'appelant. Comme toujours, un peu de recherche peut s'avérer utile ici.

L'accessoire **str** permet d'appliquer des méthodes de chaîne comme **split** directement sur des colonnes. Dans le dataset **Customer Lifetime Value**, qui contient des informations sur les clients d'une compagnie d'assurance, on peut séparer le type et le niveau de couverture à partir de la caractéristique Policy :

```python
customer = pd.read_csv("customer.csv")
```

**Decomposons**

```python
customer[["Type", "Niveau"]] = (  # Création de deux nouvelles caractéristiques
    customer["Policy"]           # à partir de la caractéristique Policy
    .str                         # via l'accessoire de chaîne
    .split(" ", expand=True)     # en séparant sur " " 
                                 # et en étendant le résultat dans des colonnes distinctes
)

# Afficher les premières lignes des données
customer[["Policy", "Type", "Niveau"]].head(10)
```

### Transformations par groupe
Les transformations par groupe permettent d'agréger des informations sur plusieurs lignes regroupées par une catégorie. Avec ces transformations, vous pouvez créer des caractéristiques comme : « le revenu moyen de l'État de résidence d'une personne » ou « la proportion de films sortis un jour de semaine, par genre ». Si vous découvrez une interaction catégorielle, une transformation par groupe basée sur cette catégorie pourrait être intéressante à explorer.

En utilisant une fonction d'agrégation, une transformation par groupe combine deux caractéristiques :
- Une caractéristique catégorielle qui définit le regroupement.
- Une autre caractéristique dont vous souhaitez agréger les valeurs.

Pour calculer le « revenu moyen par État », vous utiliseriez **State** comme caractéristique de regroupement, **mean** comme fonction d'agrégation, et **Income** comme caractéristique à agréger. En Pandas, cela se fait avec les méthodes **groupby et transform** :

```python
customer["AverageIncome"] = (
    customer.groupby("State")  # for each state
    ["Income"]                 # select the income
    .transform("mean")         # and compute its mean
)

customer[["State", "Income", "AverageIncome"]].head(10)
```
