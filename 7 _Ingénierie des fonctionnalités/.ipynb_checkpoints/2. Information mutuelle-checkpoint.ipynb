# Information mutuelle
Localisez les entités ayant le plus de potentiel.

## Introduction
Découvrir un nouveau jeu de données pour la première fois peut parfois sembler écrasant. Vous pourriez être confronté à des centaines, voire des milliers de fonctionnalités, sans même une description pour vous guider. Par où commencer ?

Une excellente première étape consiste à construire un classement avec **une métrique d’utilité des fonctionnalités**, une fonction mesurant les associations entre une fonctionnalité et la cible. Vous pourrez ensuite choisir un ensemble restreint des fonctionnalités les plus utiles pour commencer votre développement et être plus sûr que votre temps sera bien investi.

La métrique que nous utiliserons s'appelle **information mutuelle**. L'information mutuelle est très semblable à **la corrélation** en ce qu'elle mesure une relation entre deux quantités. L'avantage de l'information mutuelle est qu'elle peut détecter n'importe quel type de relation, tandis que la corrélation ne détecte que les relations linéaires.

L'information mutuelle est une excellente métrique polyvalente, particulièrement utile au début du développement des fonctionnalités, lorsque vous ne savez pas encore quel modèle utiliser. Elle est :

- facile à utiliser et à interpréter,
- efficace sur le plan computationnel,
- fondée sur des bases théoriques solides,
- résistante au surapprentissage, et
- capable de détecter n’importe quel type de relation.


---

## Information mutuelle et ce qu'elle mesure
L'information mutuelle décrit les relations en termes ***d'incertitude***. **L'information mutuelle (MI)** entre deux quantités est une mesure de l'étendue à laquelle la connaissance d'une quantité réduit l'incertitude sur l'autre. Si vous connaissiez la valeur d'une fonctionnalité, combien plus sûr seriez-vous de la cible ?

Voici un exemple issu des données de Ames Housing. Le graphique montre la relation entre la qualité extérieure d'une maison et le prix auquel elle a été vendue. Chaque point représente une maison.

**Exemple : Qualité extérieure (ExterQual) et prix de vente (SalePrice)**
Quatre catégories de qualité extérieure : médiocre, typique, bonne, excellente. Un nuage de points de prix de vente dans chaque catégorie.

- Savoir la qualité extérieure d'une maison réduit l'incertitude sur son prix de vente.

D’après le graphique, nous voyons que connaître la valeur de ExterQual devrait vous rendre plus sûr du SalePrice correspondant : chaque catégorie d’ExterQual tend à concentrer les prix de vente dans une certaine plage.

L’information mutuelle entre **ExterQual** et **SalePrice** correspond à la réduction moyenne d’incertitude sur le prix de vente calculée sur les quatre valeurs d’ExterQual. Par exemple, la catégorie "médiocre" (Fair) est moins fréquente que "typique" (Typical) et obtient donc moins de poids dans le score d’information mutuelle.

(*Note technique : Ce que nous appelons **incertitude** est mesuré à l’aide d’une quantité issue de la théorie de l’information appelée "entropie". **L’entropie** d’une variable indique approximativement : "combien de questions oui/non il faudrait poser pour décrire un événement de cette variable, en moyenne". Plus il y a de questions, plus l’incertitude est grande. L’information mutuelle correspond au nombre de questions auxquelles une fonctionnalité peut répondre sur la cible.*)



---

## Interprétation des scores d’information mutuelle
Le **score minimum** d’information mutuelle entre des quantités est de **0,0**.
- Lorsque MI est zéro, les quantités sont indépendantes : elles ne peuvent rien vous dire l’une sur l’autre.

En théorie, il n’y a pas de limite supérieure au MI.
- Cependant, dans la pratique, des valeurs supérieures à **2,0** sont rares (le MI est une quantité logarithmique, donc il augmente très lentement).

---

## À retenir lors de l’utilisation de MI
- **Évaluer le potentiel relatif** d'une fonctionnalité comme prédicteur de la cible (considérée seule).
- MI ne détecte pas les **interactions entre les fonctionnalités**. C’est une métrique univariée.
- **Utilité réelle** : Une fonctionnalité n'est utile que si sa relation avec la cible peut être apprise par le modèle. *Un score MI élevé n'assure pas que le modèle pourra exploiter cette information sans transformation préalable.*

## Exemple – Automobiles 1985
L'ensemble de données Automobile contient 193 voitures de l'année modèle 1985.

**Objectif** : prédire le prix d'une voiture (cible) à partir de 23 caractéristiques, telles que make, body_style, et horsepower.

````python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
````

````python
df = pd.read_csv("autos.csv")
df.head()
````

**Traitement des données**

1. Préparation des données :

````python
X = df.copy()
y = X.pop("price")

# Encodage des catégories
for colname in X.select_dtypes("object"):
    X[colname], _ = X[colname].factorize()

# Vérifier que toutes les fonctionnalités discrètes sont bien au format entier
discrete_features = X.dtypes == int
````

2. Calcul des scores d’information mutuelle :

````python
from sklearn.feature_selection import mutual_info_regression

def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_scores(X, y, discrete_features)
print(mi_scores)
````

3. Visualisation :

````python
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")
    plt.show()

plot_mi_scores(mi_scores)
````

La visualisation des données est un excellent complément à un classement d’utilité. Regardons de plus près quelques-unes de ces caractéristiques.

Comme on pourrait s’y attendre, la caractéristique **curb_weight** (poids à vide), qui obtient un score élevé, présente une forte relation avec le prix, la cible.

````python
sns.relplot(x="curb_weight", y="price", data=df)
plt.show()
````

La caractéristique **fuel_type** (type de carburant) a un score MI relativement bas, mais, comme on peut le voir sur la figure, elle sépare clairement deux populations de prix avec des tendances différentes en fonction de la caractéristique **horsepower** (puissance du moteur). Cela indique que **fuel_type** contribue à un ***effet d’interaction*** et pourrait ne pas être aussi négligeable qu'il n'y paraît. 

Avant de conclure qu’une caractéristique est sans importance sur la base de son score MI, il est judicieux d’explorer les éventuels **effets d’interaction** — les connaissances du domaine peuvent fournir de précieuses indications à ce sujet.

````python
sns.lmplot(x="horsepower", y="price", hue="fuel_type", data=df)
plt.show()
````

La visualisation des données est un ajout précieux à votre boîte à outils d’ingénierie des caractéristiques. Associées à des métriques d’utilité comme l’information mutuelle, des visualisations comme celles-ci peuvent vous aider à découvrir des relations importantes dans vos données. Consultez notre cours sur la visualisation des données pour en savoir plus !

# EXERCICE

Dans cet exercice, vous identifierez un ensemble initial de caractéristiques dans le dataset Ames à développer en utilisant les scores d'information mutuelle et les graphiques d'interaction.

**Code configurant l'environnement**

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.feature_selection import mutual_info_regression

# Paramètres par défaut de Matplotlib
plt.style.use("seaborn-v0_8-whitegrid")  # Correction : utilisation de la version mise à jour de seaborn
plt.rc("figure", autolayout=True)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=14,
    titlepad=10,
)

# Chargement des données
df = pd.read_csv("ames.csv")

# Fonctions utilitaires issues du tutoriel
def make_mi_scores(X, y):
    """
    Calcule les scores d'information mutuelle pour chaque caractéristique de X par rapport à y.
    """
    X = X.copy()
    # Encodage des variables catégorielles
    for colname in X.select_dtypes(["object", "category"]):
        X[colname], _ = X[colname].factorize()
    # Vérification des caractéristiques discrètes
    discrete_features = [pd.api.types.is_integer_dtype(t) for t in X.dtypes]
    # Calcul des scores d'information mutuelle
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features, random_state=0)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    # Tri des scores dans l'ordre décroissant
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

def plot_mi_scores(scores):
    """
    Trace un graphique en barres horizontales des scores d'information mutuelle.
    """
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")
````

**Comprendre l'Information Mutuelle**

Pour commencer, examinons la signification de l'information mutuelle en analysant quelques caractéristiques du dataset Ames.

````python
features = ["YearBuilt", "MoSold", "ScreenPorch"]
sns.relplot(
    x="value", 
    y="SalePrice", 
    col="variable", 
    data=df.melt(id_vars="SalePrice", value_vars=features), 
    facet_kws=dict(sharex=False),
);
````

### Question 1 : Quelle caractéristique aura le score d'information mutuelle (MI) le plus élevé avec SalePrice ?
Réponse correcte :
**YearBuilt** devrait avoir le score MI le plus élevé, car connaître l'année de construction permet de limiter SalePrice à une plage de valeurs plus restreinte.

Cela n'est généralement pas le cas pour **MoSold** (mois de vente).

Enfin, bien que **ScreenPorch** (présence de porche) fournisse un peu plus d'information que **MoSold**, cette caractéristique est souvent égale à zéro et n'apporte donc pas beaucoup d'informations globales.

#### Analyse des graphiques
- **YearBuilt** : Une relation claire est visible entre YearBuilt et SalePrice. Les maisons récentes tendent à avoir des prix plus élevés, ce qui en fait une caractéristique informative.
- **MoSold** : Le mois de vente ne semble pas influencer directement SalePrice. La distribution des prix est uniforme sur les mois.
- **ScreenPorch** : Bien que certains prix plus élevés soient associés à un porche (valeurs non nulles), la plupart des données montrent une valeur nulle pour cette caractéristique, ce qui limite sa contribution à l'information mutuelle.

### Analyse des Scores d'Information Mutuelle pour le Dataset Ames
L'objectif ici est d'identifier les caractéristiques du dataset Ames ayant le plus grand potentiel pour expliquer la variable cible SalePrice (le prix de vente).

````python
X = df.copy()
y = X.pop('SalePrice')

mi_scores = make_mi_scores(X, y)  #Calcul des scores d'information mutuelle (MI) 

#Affichage des résultats
print(mi_scores.head(20))
plot_mi_scores(mi_scores.head(20))
````

Les **20 premières caractéristiques** (celles ayant les scores MI les plus élevés) sont affichées.

### Interprétation :
1. Caractéristiques les plus importantes :
**OverallQual** (qualité globale) et **Neighborhood** (quartier) ont les scores les plus élevés, indiquant qu'elles sont fortement associées à **SalePrice**.

2. Caractéristiques supplémentaires :
Les dimensions des espaces de vie (**GrLivArea, FirstFlrSF, TotalBsmtSF**) et les attributs des garages (**GarageArea, GarageCars**) sont également des contributeurs majeurs.

3. Caractéristiques moins informatives :
Les scores d'autres variables, comme **SecondFlrSF ou FireplaceQu**, sont plus faibles mais peuvent encore être utiles dans des interactions complexes.
