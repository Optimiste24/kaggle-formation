## Importer les bibliothèques nécessaires ##

```python
# Pandas, Numpy, os et changement de répertoire de travail
import pandas as pd 
import pandas.plotting as pdl
import numpy as np 
import os


# Scikit-Learn
import sklearn as sk
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import roc_auc_score, roc_curve, auc
from sklearn.model_selection import train_test_split  ### permet de partager DA et DT
from sklearn.model_selection import StratifiedShuffleSplit   ### Fonction qui permet de faire des échantillonnage stratifié
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

# StatsModels

import statsmodels.api as sm
from statsmodels.discrete.discrete_model import Logit 

# Pour les graphiques 

%matplotlib  inline
# Ceci permet d'afficher les graphique directement dans l'output du notebbok
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
def evaluate_model(model, X_test, y_true, model_name,plot_roc=False):
    # Faire des prédictions
    y_predict = model.predict(X_test)
    y_pred = ( y_predict > 0.5).astype(int)
    
    # Calculer la matrice de confusion
    conf_matrix = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(4, 3))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Classe 0', 'Classe 1'],
                yticklabels=['Classe 0', 'Classe 1'])
    plt.xlabel('Valeurs prédites')
    plt.ylabel('Valeurs réelles')
    plt.title('Matrice de Confusion')
    plt.show()

    # Calcul des métriques d'évaluation
    accuracy = round(accuracy_score(y_true, y_pred),2)
    precision = round(precision_score(y_true, y_pred),2)
    recall = round(recall_score(y_true, y_pred),2)
    

    # Affichage des métriques
    print('\033[1m',"Métriques d'évaluation pour le modèle {} :".format(model_name), '\033[0m')
    print("Exactitude:", accuracy)
    print("Précision:", round(precision, 3))
    print("Rappel:", recall)
    

    # Tracer la courbe ROC
    
    fpr, tpr, _ = roc_curve(y_true, y_predict)
    auc_score = auc(fpr, tpr)
    print("L'AUC du modèle {} est : {:.4f}".format(model_name, auc_score))
    
    if plot_roc:
        plt.figure(figsize=(10, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2, label='{} ROC curve (area = {:.2f})'.format(model_name, auc_score))
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlabel('Taux de faux positif')
        plt.ylabel('Taux de vrai positif')
        plt.title('Courbe ROC - {}'.format(model_name))
        plt.legend(loc='lower right')
        plt.show()
```

### Nous allons partagé les données, mais en faisant attention que le pourcentage  des classes restent presque identique ####

<div class="alert alert-success">

Il y a une fonction pour faire cela :  `StratifiedShuffleSplit `  et une  méthode  `split() `

```python
from sklearn.model_selection import StratifiedShuffleSplit


strat_split = StratifiedShuffleSplit(n_splits=1, test_size=0.3, random_state=452)


for train_index, test_index in strat_split.split(X, y):
    X_train_strat, X_test_strat = X.iloc[train_index], X.iloc[test_index]
    y_train_strat, y_test_strat = y.iloc[train_index], y.iloc[test_index]

```

## **<font color="#800080">5. Les arbres, les graphiques de la longueur de l’arbre et le taux de bonnes réponses selon le paramètre Alpha : </font>** ##

```python
bq.columns
```

```python
bq2 = pd.get_dummies(bq, columns=(['Compt', 'Famil']))
bq2[0:3]
```

```python
## J'enlève des colonnes pour mes variables explicatives, je transforme True et False en chiffre 0 et 1 et j'ajoute une constante. 

X2 = bq2.drop(columns=['Age','Exper','Educa','Compt_Non','Pret_Pers'], axis=1)
X2["Compt_Oui"]=X2["Compt_Oui"].astype(int)
X2["Famil_1"]=X2["Famil_1"].astype(int)
X2["Famil_2"]=X2["Famil_2"].astype(int)
X2["Famil_3"]=X2["Famil_3"].astype(int)
X2["Famil_4"]=X2["Famil_4"].astype(int)

y2 = bq2['Pret_Pers']=='Oui'
y2=y2.astype(int)


X2.head(), y2.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score



# Division des données en ensemble d'entraînement et ensemble de test
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=42)

# Paramètres alpha à tester
alphas = np.linspace(0.0, 0.005, 10)

# Initialisation des listes pour stocker les résultats
tree_lengths = []
accuracies = []

# Boucle sur les différents alpha
for alpha in alphas:
    # Création de l'arbre avec le paramètre alpha donné
    tree = DecisionTreeClassifier(random_state=42, ccp_alpha=alpha)
    
    # Entraînement de l'arbre
    tree.fit(X2_train, y2_train)
    
    # Longueur de l'arbre
    tree_length = tree.tree_.node_count
    tree_lengths.append(tree_length)
    
    # Prédiction sur l'ensemble de test
    y2_pred = tree.predict(X2_test)
    
    # Taux de bonnes réponses
    accuracy = accuracy_score(y2_test, y2_pred)
    accuracies.append(accuracy)

# Tracé des résultats
plt.figure(figsize=(12, 6))

# Graphique de la longueur de l'arbre en fonction du paramètre alpha
plt.subplot(1, 2, 1)
plt.plot(alphas, tree_lengths, marker='o')
plt.xlabel('Alpha')
plt.ylabel('Longueur de l\'arbre')
plt.title('Longueur de l\'arbre en fonction d\'Alpha')

# Graphique du taux de bonnes réponses en fonction du paramètre alpha
plt.subplot(1, 2, 2)
plt.plot(alphas, accuracies, marker='o', color='orange')
plt.xlabel('Alpha')
plt.ylabel('Taux de bonnes réponses')
plt.title('Taux de bonnes réponses en fonction d\'Alpha')

plt.tight_layout()
plt.show()
```

## **<font color="#800080">5.Choix de la valeur d'Alpha et l'influence des variables sur graphe : </font>** ##

```python
# la valeur d'alpha 
alpha_choisi = 0.001

# Créer l'arbre avec la valeur d'alpha choisie
arbre_choisi = DecisionTreeClassifier(random_state=42, ccp_alpha=alpha_choisi)
arbre_choisi.fit(X2_train, y2_train)

# Afficher l'arbre
plt.figure(figsize=(16, 8))
plot_tree(arbre_choisi, filled=True, feature_names=X2.columns.tolist(), class_names=['Non Prêt', 'Prêt'], rounded=True)
plt.show()
```

## Influence des variables

```python
# Obtenir l'importance des variables
importances = arbre_choisi.feature_importances_

# Trier les indices des variables par importance décroissante
indices = np.argsort(importances)[::-1]

# Tracer l'importance des variables
plt.figure(figsize=(10, 6))
plt.bar(range(X2.shape[1]), importances[indices], align='center')
plt.xticks(range(X2.shape[1]), X2.columns[indices], rotation=45)
plt.xlabel('Variables')
plt.ylabel('Importance')
plt.title('Importance des variables dans la prise de décision')
plt.show()

```

## **<font color="#800080">6. Arbre de décision avec des variables pondérées : </font>** ##

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
import numpy as np

# Définir les poids des échantillons (exemple de pondération égale pour tous les échantillons)
sample_weights = np.ones(len(X2_train))

# Choisir la valeur d'alpha
alpha_choisi = 0.001

# Créer l'arbre avec la valeur d'alpha choisie et les poids des échantillons
arbre_pondere = DecisionTreeClassifier(random_state=42, ccp_alpha=alpha_choisi)
arbre_pondere.fit(X2_train, y2_train, sample_weight=sample_weights)

# Convertir l'Index en liste
feature_names_list = list(X2.columns)

# Afficher l'arbre avec la liste convertie
plt.figure(figsize=(16, 8))
plot_tree(arbre_pondere, filled=True, feature_names=feature_names_list, class_names=['Non Prêt', 'Prêt'], rounded=True)
plt.show()
```

## **<font color="#800080">7. Matrice de Confusion : </font>** ##

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, roc_auc_score

# Prédictions sur l'ensemble de test
y_pred = arbre_choisi.predict(X2_test)

# Matrice de confusion sans pondération
conf_matrix_sans_pond = confusion_matrix(y2_test, y_pred)
print("Matrice de Confusion sans pondération :\n", conf_matrix_sans_pond)

# Calcul des métriques sans pondération
accuracy_sans_pond = accuracy_score(y2_test, y_pred)
precision_sans_pond = precision_score(y2_test, y_pred)
recall_sans_pond = recall_score(y2_test, y_pred)
auc_sans_pond = roc_auc_score(y2_test, y_pred)

print("\nExactitude sans pondération:", accuracy_sans_pond)
print("Précision sans pondération:", precision_sans_pond)
print("Rappel sans pondération:", recall_sans_pond)
print("AUC sans pondération:", auc_sans_pond)

# Prédictions sur l'ensemble de test avec pondération
y_pred_pondere = arbre_pondere.predict(X2_test)

# Matrice de confusion avec pondération
conf_matrix_avec_pond = confusion_matrix(y2_test, y_pred_pondere)
print("\nMatrice de Confusion avec pondération :\n", conf_matrix_avec_pond)

# Calcul des métriques avec pondération
accuracy_avec_pond = accuracy_score(y2_test, y_pred_pondere)
precision_avec_pond = precision_score(y2_test, y_pred_pondere)
recall_avec_pond = recall_score(y2_test, y_pred_pondere)
auc_avec_pond = roc_auc_score(y2_test, y_pred_pondere)

print("\nExactitude avec pondération:", accuracy_avec_pond)
print("Précision avec pondération:", precision_avec_pond)
print("Rappel avec pondération:", recall_avec_pond)
print("AUC avec pondération:", auc_avec_pond)
```

## **<font color="#800080">8. Courbe ROC : </font>** ##

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Probabilités des classes sans pondération
probas_sans_pond = arbre_choisi.predict_proba(X2_test)[:, 1]

# Probabilités des classes avec pondération
probas_avec_pond = arbre_pondere.predict_proba(X2_test)[:, 1]

# Courbe ROC sans pondération
fpr_sans_pond, tpr_sans_pond, _ = roc_curve(y2_test, probas_sans_pond)
roc_auc_sans_pond = auc(fpr_sans_pond, tpr_sans_pond)

# Courbe ROC avec pondération
fpr_avec_pond, tpr_avec_pond, _ = roc_curve(y2_test, probas_avec_pond)
roc_auc_avec_pond = auc(fpr_avec_pond, tpr_avec_pond)

# Afficher les courbes ROC
plt.figure(figsize=(8, 8))
plt.plot(fpr_sans_pond, tpr_sans_pond, label='Sans pondération (AUC = {:.2f})'.format(roc_auc_sans_pond))
plt.plot(fpr_avec_pond, tpr_avec_pond, label='Avec pondération (AUC = {:.2f})'.format(roc_auc_avec_pond))
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Aléatoire')
plt.xlabel('Taux de Faux Positifs')
plt.ylabel('Taux de Vrais Positifs')
plt.title('Courbe ROC')
plt.legend()
plt.show()
```

## **<font color="#800080"> Perspective de recherche : BOOSTING / SVM  </font>** ##


BOOSTING : Amélioration de la performance des modèles en mettant l'accent sur les erreurs de prédictions précédentes.

SVM : Les SVM permettent d'assigner des poids différents aux échantillons, ce qui peut être utile lorsque les classes sont déséquilibrées. Les échantillons de la classe minoritaire peuvent recevoir un poids plus élevé pour donner plus d'importance à ces échantillons lors de l'apprentissage du modèle.

## *BOOSTING* ##

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


# Supposons que X2 et y2 sont les données d'entraînement
X3_train, X3_test, y3_train, y3_test = train_test_split(X2, y2, test_size=0.2, random_state=42)

# Créez le modèle Gradient Boosting
boosted_model = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, random_state=42)

# Entraînez le modèle sur les données d'entraînement
boosted_model.fit(X3_train, y3_train)

# Prédictions sur les données de test
y3_pred = boosted_model.predict(X3_test)

# Évaluez la précision
accuracy = accuracy_score(y3_test, y3_pred)
print("Accuracy:", accuracy), boosted_model, y3_pred
```

Commentaire : Un score d'exactitude de 0.9830 (ou 98.30%) est généralement très élevé, ce qui suggère que le modèle de boosting que vous avez appliqué fonctionne bien sur vos données de prêts personnels.

## *SVM* ##

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

# Supposez que X4 et y4 soient les données d'entraînement et de cible

# Division des données en ensemble d'entraînement et ensemble de test
X4_train, X4_test, y4_train, y4_test = train_test_split(X2, y2, test_size=0.2, random_state=42)

# Définir les poids pour chaque classe (ajustez-les selon votre besoin)
weights = {0: 1, 1: 5}  # Par exemple, poids 5 pour la classe 1 (prêt) et poids 1 pour la classe 0 (non prêt)

# Créer le modèle SVM avec poids
svm_model = svm.SVC(class_weight=weights)

# Entraîner le modèle
svm_model.fit(X4_train, y4_train)

# Faire des prédictions sur l'ensemble de test
y4_pred = svm_model.predict(X4_test)

# Mesurer la performance du modèle
accuracy = accuracy_score(y4_test, y4_pred)
conf_matrix = confusion_matrix(y4_test, y4_pred)

print("Accuracy:", accuracy)
print("Confusion Matrix:\n", conf_matrix)
```

Commentaire : Le résultat de votre modèle SVM sur les données de prêts personnels indique une accuracy de 95.03%, ce qui est relativement élevé.
