### Choisir un jeu de données
Dans ce tutoriel, nous travaillerons avec un jeu de données du département américain des transports qui suit les retards de vols.

En ouvrant ce fichier CSV dans Excel, vous verrez :

une ligne pour chaque mois (1 = janvier, 2 = février, etc.),
une colonne pour chaque code de compagnie aérienne.
Exemple :
AA	AS	B6	...
1 (janvier)	6.95 min	-0.32 min	7.34 min	...
Chaque valeur représente le retard moyen d'arrivée (en minutes) pour une compagnie aérienne donnée et un mois précis (en 2015).

Les valeurs négatives indiquent des vols qui arrivaient en avance.

**Exemple** : en janvier, les vols d’American Airlines (code : AA) arrivaient avec environ 7 minutes de retard, tandis que ceux d’Alaska Airlines (code : AS) arrivaient avec environ 3 minutes d’avance en avril.

---

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Chemin du fichier à lire
flight_filepath = "flight_delays.csv"

# Charger le fichier dans une variable flight_data
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

Le code est légèrement plus court que dans le tutoriel précédent, car ici les étiquettes des lignes (colonne "Month") ne sont pas des dates, donc nous n’ajoutons pas parse_dates=True.
Cependant, nous précisons toujours :

- Le chemin du fichier (flight_filepath),
- Le nom de la colonne qui sera utilisée pour indexer les lignes (index_col="Month").

```python
# Afficher les données
flight_data
```

---

## Graphiques à barres
Supposons que nous souhaitons créer un graphique à barres montrant le retard moyen d’arrivée pour Spirit Airlines (code : NK) par mois :

```python
plt.clf() # effacé tout graphique précédent

# Définir la largeur et la hauteur du graphique
plt.figure(figsize=(10,6))

# Ajouter un titre
plt.title("Retard moyen d'arrivée des vols Spirit Airlines, par mois")

# Créer un graphique à barres
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Ajouter une étiquette pour l'axe vertical
plt.ylabel("Retard à l'arrivée (en minutes)")

# Forcer l'affichage du graphique
plt.show()
```

---

## Cartes thermiques (heatmaps)
Nous pouvons également utiliser une carte thermique pour visualiser rapidement les tendances dans les données :

```python
# Définir la largeur et la hauteur du graphique
plt.figure(figsize=(14,7))

# Ajouter un titre
plt.title("Retard moyen d'arrivée pour chaque compagnie aérienne, par mois")

# Créer une carte thermique
sns.heatmap(data=flight_data, annot=True)

# Ajouter une étiquette pour l'axe horizontal
plt.xlabel("Compagnie aérienne")

# l'affichage du graphique
plt.show()
```

# Exercice : Analyse des Critiques de Jeux Vidéo
Dans cet exercice, vous utiliserez vos nouvelles connaissances pour proposer une solution à un scénario réel. Vous devrez importer des données dans Python, répondre à des questions à partir des données, et générer des **graphiques en barres** ainsi que des **cartes thermiques** (heatmaps) pour comprendre les motifs présents dans les données.

#### Scénario
Vous avez récemment décidé de créer votre propre jeu vidéo ! En tant que lecteur assidu des critiques de jeux vidéo d'IGN, vous êtes au courant des dernières sorties, ainsi que des notes attribuées par les experts. Ces notes varient de 0 (Catastrophe) à 10 (Chef-d'œuvre).

Vous souhaitez utiliser les critiques d'IGN pour guider la conception de votre futur jeu. Heureusement, quelqu'un a résumé ces notes dans un fichier CSV très utile, que vous pouvez exploiter pour vos analyses.



```python
# Chemin d'accès au fichier à lire
ign_filepath = "ign_scores.csv"

# Lire le fichier dans une variable ign_data et définir "Platform" comme index
ign_data = pd.read_csv(ign_filepath, index_col="Platform")

# Afficher l'ensemble des données
print(ign_data)
```

```python
# Quelle est la note moyenne la plus élevée reçue par les jeux PC pour un genre ?
high_score = ign_data.loc["PC"].max()

# Sur la plateforme Playstation Vita, quel genre a reçu la note moyenne la plus basse ?
worst_genre = ign_data.loc["PlayStation Vita"].idxmin()

print(high_score, worst_genre)
```

###  Graphique à barres pour les scores moyens des jeux de course

```python
# Définir la largeur et la hauteur de la figure
plt.figure(figsize=(8, 6))

# Créer un graphique à barres montrant le score moyen des jeux de course par plateforme
sns.barplot(x=ign_data['Racing'], y=ign_data.index, orient='h', palette="viridis")

# Ajouter un titre
plt.title("Score moyen des jeux de course par plateforme", fontsize=14)

# Ajouter une étiquette pour l'axe horizontal
plt.xlabel("Score moyen", fontsize=12)

# Ajouter une étiquette pour l'axe vertical
plt.ylabel("Platform", fontsize=12)

# Afficher le graphique
plt.show()
```

## Carte thermique pour les scores moyens des jeux de course

```python
# Définir la taille du graphique
plt.figure(figsize=(10, 10))

# Créer la heatmap
sns.heatmap(ign_data, annot=True)

# Ajouter un titre
plt.title("Average Game Score by Genre and Platform", fontsize=16)

# Afficher la heatmap
plt.show()
```

**Solution** : Les jeux de simulation sur Playstation 4 reçoivent les meilleures notes moyennes (9,2). Les jeux de tir et de combat sur Game Boy Color reçoivent les pires notes moyennes (4,5).
