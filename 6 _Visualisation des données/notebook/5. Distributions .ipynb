# Distributions
Création d’histogrammes et de diagrammes de densité

Introduction
Dans ce tutoriel, nous allons explorer **les histogrammes** et les **diagrammes de densité**.

### Sélection d’un jeu de données
Nous travaillerons avec un ensemble de données contenant 150 fleurs, réparties également entre trois espèces différentes d’iris : **Iris setosa, Iris versicolor et Iris virginica**.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

### Chargement et examen des données
Chaque ligne du jeu de données représente une fleur, avec quatre mesures :
- La longueur et la largeur du sépale,
- La longueur et la largeur du pétale.

Les données incluent également l’espèce correspondante.

```python
# Chemin du fichier à lire
iris_filepath = "iris.csv"

# Lecture des données
iris_data = pd.read_csv(iris_filepath, index_col="Id")

# Affichage des 5 premières lignes
iris_data.head()
```

---

# Histogrammes
Pour observer comment la longueur des pétales varie parmi les iris, utilisons la commande ***sns.histplot*** :

```python
# Création de l'histogramme
sns.histplot(iris_data['Petal Length (cm)'])
plt.show()
```

---

## Diagrammes de densité (KDE Plots)
Un KDE plot (estimation par noyau) est une version lissée d’un histogramme. Utilisez ***sns.kdeplot*** avec shade=True pour colorer la zone sous la courbe :

```python
# Diagramme de densité (KDE)
sns.kdeplot(data=iris_data['Petal Length (cm)'], shade=True)
plt.show()
```

---

### Diagrammes de densité 2D
Nous pouvons aussi créer un KDE plot en deux dimensions à l’aide de ***sns.jointplot*** :

```python
# Diagramme de densité 2D
sns.jointplot(x=iris_data['Petal Length (cm)'], y=iris_data['Sepal Width (cm)'], kind="kde")
plt.show()
```

- La partie centrale du graphique représente la densité combinée des deux variables.
- Les courbes en haut et à droite sont des KDE plots individuels pour chaque axe.


---

### Diagrammes avec code couleur
Nous pouvons colorer les histogrammes et KDE plots par espèce :

**Histogrammes pour chaque espèce**

```python
sns.histplot(data=iris_data, x='Petal Length (cm)', hue='Species')

# Ajout du titre
plt.title("Histogram of Petal Lengths, by Species")
plt.show()
```

**KDE plots pour chaque espèce**

```python
sns.kdeplot(data=iris_data, x='Petal Length (cm)', hue='Species', shade=True)

# Ajout du titre
plt.title("Distribution of Petal Lengths, by Species")
plt.show()
```

**Observation intéressante** :

Les espèces semblent se regrouper en deux catégories :
1. Iris versicolor et Iris virginica ont des valeurs similaires pour la longueur des pétales.
2. Iris setosa se distingue nettement avec des longueurs de pétales inférieures à 2 cm.

Chaque tumeur est classée comme **bénigne** (non cancéreuse) ou **maligne** (cancéreuse).

#### Étape 1 : Charger les données
Dans cette étape, vous allez charger les données.
- Chargez le fichier de données dans un DataFrame nommé cancer_data.
- Le chemin correspondant est cancer_filepath.

```python
# Chemin d'accès au fichier à lire
cancer_filepath = "cancer.csv"

# Complétez la ligne ci-dessous pour lire le fichier dans une variable cancer_data
cancer_data = pd.read_csv(cancer_filepath, index_col="Id")

# Visualisez les premières lignes des données
cancer_data.head()
```

#### étape 2 : Examiner les données

1. Affichez les 5 premières lignes des données pour mieux comprendre le contenu du dataset.
2. Chaque ligne correspond à une image différente, avec :
- Une colonne Diagnosis qui classe les tumeurs comme bénignes (B) ou malignes (M).
- 30 colonnes contenant différentes mesures issues des images.

**Questions** :
- Quelle est la plus grande valeur de Perimeter (mean) dans les 5 premières lignes ?
- Quelle est la valeur de Radius (mean) pour la tumeur avec l'Id 8510824 ?

```python
# Remplissez la ligne ci-dessous : Quelle est la plus grande valeur de 'Perimeter (mean)' dans les 5 premières lignes ?
max_perim = cancer_data['Perimeter (mean)'].head().max()

# Remplissez la ligne ci-dessous : Quelle est la valeur de 'Radius (mean)' pour la tumeur avec Id 8510824 ?
mean_radius = cancer_data.loc[8510824, 'Radius (mean)']

print(max_perim, mean_radius)
```

#### étape 3, 
Partie A, afin de créer des histogrammes séparés pour les tumeurs bénignes et malignes en utilisant une seule figure.

```python
# Créer des histogrammes pour les tumeurs bénignes et malignes
plt.figure(figsize=(10, 6))  # Définir la taille de la figure

sns.histplot(data=cancer_data, x='Area (mean)', hue='Diagnosis')

# Ajouter un titre et des légendes
plt.title('Distribution of Area (mean) for Benign and Malignant Tumors', fontsize=16)

# Afficher le graphique
plt.show()
```

### Solution Explication :

1. **Malignant Tumors Have Higher Values**:
   - Sur les histogrammes, on observe que les valeurs moyennes de **`Area (mean)`** pour les tumeurs malignes se situent nettement à droite de celles des tumeurs bénignes. Cela indique que les tumeurs malignes ont des **valeurs moyennes d'aire plus élevées**.

2. **Malignant Tumors Have a Larger Range**:
   - Les tumeurs malignes couvrent une **plage de valeurs plus large** pour **`Area (mean)`**, allant de valeurs proches de celles des tumeurs bénignes jusqu'à des valeurs beaucoup plus élevées. En revanche, les tumeurs bénignes ont une plage de valeurs plus restreinte.

Ces observations sont cohérentes avec le fait que les tumeurs malignes ont tendance à être plus grandes et à présenter des caractéristiques plus variables, ce qui peut aider à leur classification.

---

**créer les deux tracés KDE pour Radius (worst), séparant les tumeurs bénignes et malignes.**

```python
# KDE plots for benign and malignant tumors
sns.kdeplot(data=cancer_data, 
            x='Radius (worst)', 
            hue='Diagnosis', 
            shade=True, 
            alpha=0.5)

# Add title and labels
plt.title('Distribution of Radius (worst), by Diagnosis')
plt.xlabel('Radius (worst)')
plt.ylabel('Density')

# Show the plot
plt.show()
```

### Explication de la solution :

Le graphique KDE montre la densité des valeurs de la colonne **`Radius (worst)`** pour les tumeurs bénignes et malignes. Autour de la valeur 25, la courbe correspondant aux **tumeurs malignes** est nettement plus élevée que celle des **tumeurs bénignes**. Cela indique que les tumeurs ayant une valeur de **`Radius (worst)`** égale à 25 sont beaucoup plus fréquentes parmi les cas malins.

Un algorithme précis apprendrait ce type de schéma à partir des données et utiliserait cette information pour classer une tumeur avec cette mesure comme maligne.

**Cette logique illustre le fonctionnement des modèles d'apprentissage automatique, où les modèles se basent sur les tendances observées dans les données pour faire des prédictions.**
