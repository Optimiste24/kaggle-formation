# À propos de Dataset
Cet ensemble de données contient **divers indicateurs de santé et facteurs de risque liés aux maladies cardiaques**. Des paramètres tels que l’âge, le sexe, la tension artérielle, le taux de cholestérol, les habitudes tabagiques et les habitudes d’exercice ont été recueillis pour analyser le risque de maladie cardiaque et contribuer à la recherche en santé. L’ensemble de données peut être utilisé par les professionnels de la santé, les chercheurs et les analystes de données pour examiner les tendances liées aux maladies cardiaques, identifier les facteurs de risque et effectuer diverses analyses liées à la santé.

Colonnes: 

- **Âge** : L’âge de l’individu.
- **Sexe** : Le sexe de l’individu (masculin ou féminin).
- **Pression artérielle** : La pression artérielle de l’individu (systolique).
- **Taux de cholestérol** : Le taux de cholestérol total de l’individu.
- **Habitudes d’exercice** : Les habitudes d’exercice de l’individu (faible, moyenne, élevée).
- **Tabagisme** : Que la personne fume ou non (Oui ou Non).
- **Maladie cardiaque familiale** : S’il y a des antécédents familiaux de maladie cardiaque (oui ou non).
- **Diabète** : Si la personne est atteinte de diabète (Oui ou Non).
- **IMC** : L’indice de masse corporelle de l’individu.
- **Hypertension artérielle** : Si la personne souffre d’hypertension artérielle (Oui ou Non).
- **Cholestérol HDL bas** : Si la personne a un faible cholestérol HDL (Oui ou Non).
- **Cholestérol LDL élevé** : Si la personne a un taux élevé de cholestérol LDL (Oui ou Non).
- **Consommation d’alcool** : Le niveau de consommation d’alcool de l’individu (aucun, faible, moyen, élevé).
- **Niveau de stress** : Le niveau de stress de l’individu (faible, moyen, élevé).
- **Heures de sommeil** : Le nombre d’heures de sommeil de l’individu.
- **Consommation de sucre** : Le niveau de consommation de sucre de l’individu (faible, moyen, élevé).
- **Taux de triglycérides** : Le taux de triglycérides de l’individu.
- **Glycémie à jeun** : La glycémie à jeun de l’individu.
- **Niveau de CRP** : Le niveau de protéine C-réactive (un marqueur de l’inflammation).
- **Niveau d’homocystéine** : Le taux d’homocystéine de l’individu (un acide aminé qui affecte la santé des vaisseaux sanguins).
- **Maladie du cœur** : L’état de la maladie du cœur de la personne (Oui ou Non).

## 1. Tendances : Suivre l'évolution des indicateurs de santé avec l'âge
Graphique utilisé : *sns.lineplot*

L'objectif ici est d'explorer comment la **taux de cholesterol** varie en fonction de **l'âge**, et de comparer cette relation entre les **hommes et les femmes**.

```python
print(my_data.columns)
```

```python
# Configurer la taille de la figure
plt.figure(figsize=(12, 6))

# Tracer les tendances pour le taux cholestérol  par âge
sns.lineplot(data=my_data, x="Age", y="Cholesterol Level", hue="Gender")

# Afficher le graphique
plt.show()
```
