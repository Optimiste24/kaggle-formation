### Sélection d’un jeu de données
Le dataset utilisé suit les streams journaliers globaux de chansons populaires sur **Spotify** entre 2017 et 2018 :

- "Shape of You" (Ed Sheeran)
- "Despacito" (Luis Fonzi)
- "Something Just Like This" (The Chainsmokers & Coldplay)
- "HUMBLE." (Kendrick Lamar)
- "Unforgettable" (French Montana)

Exemple de données (les premières lignes) :

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Chemin vers le fichier CSV
spotify_filepath = "spotify.csv"

# Chargement des données
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)

# Affichage des 5 premières lignes
spotify_data.head()
```

---

### Visualisation des données
Pour créer un graphique linéaire simple, utilisez :

```python
# Conversion des données en format long
spotify_long = spotify_data.reset_index().melt(id_vars="Date", var_name="Song", value_name="Streams")

# Tracé des courbes
sns.lineplot(data=spotify_long, x="Date", y="Streams", hue="Song")
plt.title("Nombre de streams pour différentes chansons")
plt.xlabel("Date")
plt.ylabel("Nombre de streams")
plt.legend(title="Chanson")
plt.show()
```
