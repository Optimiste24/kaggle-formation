### Choix des types de tracé et des styles personnalisés

Personnalisez vos graphiques et donnez-leur un aspect élégant.

Dans ce cours, vous avez appris à créer de nombreux types de graphiques différents. Maintenant, vous allez organiser vos connaissances avant d'apprendre quelques commandes rapides que vous pouvez utiliser pour changer le style de vos graphiques.

#### Ce que vous avez appris :

Il n'est pas toujours facile de décider quel type de graphique est le meilleur pour raconter l’histoire de vos données. C’est pourquoi nous avons divisé les types de graphiques en trois grandes catégories pour vous aider dans cette tâche.

1. **Tendances** - Une tendance est définie comme un modèle de changement.
   - **sns.lineplot** : Les graphiques en ligne sont idéaux pour montrer les tendances sur une période donnée. Plusieurs lignes peuvent être utilisées pour montrer les tendances dans plus d'un groupe.

2. **Relations** - Il existe de nombreux types de graphiques que vous pouvez utiliser pour comprendre les relations entre les variables de vos données.
   - **sns.barplot** : Les graphiques à barres sont utiles pour comparer des quantités correspondant à différents groupes.
   - **sns.heatmap** : Les cartes thermiques peuvent être utilisées pour détecter des motifs codés par couleur dans des tableaux de nombres.
   - **sns.scatterplot** : Les nuages de points montrent la relation entre deux variables continues ; si elles sont colorées, on peut aussi afficher la relation avec une troisième variable catégorielle.
   - **sns.regplot** : Inclure une ligne de régression dans un nuage de points permet de voir plus facilement une relation linéaire entre deux variables.
   - **sns.lmplot** : Cette commande est utile pour tracer plusieurs lignes de régression si le nuage de points contient plusieurs groupes codés par couleur.
   - **sns.swarmplot** : Les nuages de points catégoriels montrent la relation entre une variable continue et une variable catégorielle.

3. **Distribution** - On visualise les distributions pour montrer les valeurs possibles que l’on peut s'attendre à voir dans une variable, ainsi que leur probabilité.
   - **sns.histplot** : Les histogrammes montrent la distribution d'une seule variable numérique.
   - **sns.kdeplot** : Les graphiques KDE (ou graphiques KDE 2D) montrent une estimation, lisse, de la distribution d'une seule variable numérique (ou de deux variables numériques).
   - **sns.jointplot** : Cette commande est utile pour afficher simultanément un graphique KDE 2D avec les graphiques KDE correspondants pour chaque variable individuelle.

### Changer le style avec Seaborn

Toutes les commandes ont fourni un joli style par défaut à chaque graphique. Cependant, il peut être utile de personnaliser l’apparence de vos graphiques, et heureusement, cela peut être accompli en ajoutant simplement une ligne de code supplémentaire !

Comme toujours, nous devons commencer par configurer l'environnement de codage. (Ce code est caché, mais vous pouvez le déverrouiller en cliquant sur le bouton "Code" juste en dessous de ce texte, à droite.)

Nous allons travailler avec le même code que nous avons utilisé pour créer un graphique linéaire dans un tutoriel précédent. Le code ci-dessous charge le jeu de données et crée le graphique.

```python
# Chemin du fichier à lire
spotify_filepath = "../input/spotify.csv"

# Lire le fichier dans une variable spotify_data
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)

# Graphique linéaire
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
<AxesSubplot:xlabel='Date'>
```

Nous pouvons rapidement changer le style de la figure en utilisant un thème différent avec une seule ligne de code.

```python
# Changer le style de la figure en "dark"
sns.set_style("dark")

# Graphique linéaire
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
<AxesSubplot:xlabel='Date'>
```

Seaborn propose cinq thèmes différents : 
1. "darkgrid" 
2. "whitegrid" 
3. "dark" 
4. "white" 
5. "ticks"

Il vous suffit d'utiliser une commande similaire à celle du code ci-dessus (avec le thème choisi rempli) pour le changer.

# Exercice : Exploration des différents styles de graphiques
Dans cet exercice, vous allez explorer différents styles de graphiques pour voir quelles combinaisons de couleurs et polices vous préférez !

#### Configuration
Exécutez la cellule suivante pour importer et configurer les bibliothèques Python nécessaires pour compléter l'exercice.

````python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Configuration terminée")

# Chemin du fichier à lire
spotify_filepath = "spotify.csv"

# Lire le fichier dans une variable spotify_data
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)
spotify_data.head()
````

#### Essayer les styles de Seaborn
Exécutez la commande suivante pour essayer le thème **"dark"**.


````python
# Changer le style de la figure
sns.set_style("dark")

# Graphique linéaire
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
plt.show()
````

**Thème : "darkgrid"**.

````python
# Changer le style de la figure
sns.set_style("darkgrid")

# Graphique linéaire
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
plt.show()
````

**Thème : "whitegrid".**

````python
# Changer le style de la figure
sns.set_style("whitegrid")

# Graphique linéaire
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
plt.show()
````

**Thème : "ticks".**

````python
# Changer le style de la figure
sns.set_style("ticks")

# Graphique linéaire
plt.figure(figsize=(12,6))
sns.lineplot(data=spotify_data)
plt.show()
````
