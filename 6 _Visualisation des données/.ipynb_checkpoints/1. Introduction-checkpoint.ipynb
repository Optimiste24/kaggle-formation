**Configuration minimale à inclure au début de chaque notebook :**

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

---
Chargement des données

**Exemple : Classement FIFA historique**


```python
fifa_filepath = "fifa.csv"
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
fifa_data.head()
```

---

### Création d'un graphique simple

Exemple : Diagramme linéaire des classements FIFA


```python
plt.figure(figsize=(16,6))  # Définit la taille du graphique
sns.lineplot(data=fifa_data)  # Crée un diagramme linéaire montrant l'évolution des classements FIFA
```

```python
# Création du graphique
plt.figure(figsize=(16, 6))  # Définit la taille du graphique

# Tracer une ligne pour chaque pays
for country in fifa_data.columns:
    sns.lineplot(data=fifa_data, x=fifa_data.index, y=country, label=country)

# Ajouter des titres et des étiquettes
plt.title("Évolution des classements FIFA par pays", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Classement FIFA", fontsize=12)
plt.legend(title="Pays", fontsize=10)
plt.xticks(rotation=45)  # Rotation des dates pour améliorer la lisibilité
plt.grid(True)  # Ajouter une grille pour faciliter la lecture

# Afficher le graphique
plt.show()
```
